### Integration methods

Just as there are rules for differentiation, there are rules for integration.

A universally-applicable rule is for any indefinite integral, we must always include the arbitrary constant of $ + C $

At RMIT, the Calc1 lecturer had this to say as a checklist for solving an integral problem:

1. Is it simple?
2. Can you use substitution?
3. Can it be done by parts?
4. Can you find a formula in a table? (People have solved many hard problems before and recorded them in tables. Give values names so that you can recognise when a formula is a appropriate).
5. Change how it looks. Make it cosmetically different. Write terms in different forms from their definitions. This may make you realise it is solveable in some other way. Once it looks different, try steps 1-3 again.

### Integration by substitution

This is the opposite of the chain rule.

If $ y = f(x) \cdot x' $

Then $ \int (y) dx = g(x) + C $

Eg:

$ y = (x^2 + 1)^5 2x $

There is a function inside a function here. Let the inner function be:

$ u = x^2 + 1 $

For this method to work, we need the derivative $ \frac{du}{dx} = 2x $. This needs to be substituted like this:

$ \int(u^5 \frac{du}{dx}) dx $

If we remember the definition for the chain rule:

$ y = f(u(x)) $

$ \frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} $

Then finding the antiderivative of $ \frac{dy}{du} $ will complete the process.

$ \int(\frac{dy}{du} \cdot \frac{du}{dx}) dx = \frac{1}{6} u^6 + C$

Substitute in the value of $ u $:

$ \frac{1}{6} (x^2 + 1)^6 + C $

The result above is the solution; differentiating would find:

$ (x^2 + 1)^5 2x $

#### When the  $\frac{du}{dx}$ is a multiple

Consider:

$ \int[ (1-x^3)^7 x^2 ]dx $

Then:

$ u = 1-x^3 $

But:

$\frac{du}{dx} = -3x^2 $, which doesn't match $ x^2 $.

This is still ok because one is a multiple of the other.

$ \frac{-1}{3} \cdot \frac{du}{dx} = x^2 $

Then:

$ \int[ u^7 \cdot \frac{-1}{3} \cdot \frac{du}{dx} ]dx$

Because integration is a linear operator:

$ \frac{-1}{3} \cdot \int[ u^7 \cdot \frac{du}{dx} ]dx$

$ \frac{-1}{3} \cdot \frac{1}{8} u^8 + C$

Substitute the vale of $ u $

$ \left(\frac{-(1-x^3)}{24}\right)^8 + C$


#### Definite integrals with substitution

This is yet another rule for using substitution.

Let's consider:

$ \int_{-1}^1 (3x^2 \sqrt{x^3 + 1}) dx $

$ u = x^3 +1 $

$ \frac{du}{dx} = 3x^2 $

#### Method 1

In the above methods, we're making $ x $ disappear from the function that is being integrated and replacing it with a function, $ u $. Therefore if the function being integrated has a definite integral for $x$. When we replace $ x $ with $ u $ then we should evaluate the function $ u $ at those values.

Let's replace the intervals:

$ u(-1) = 0 $

$ u(i) = 2 $

Then expressed in terms of the function $ u $:

$ \int_{0}^2 (\frac{du}{dx} \sqrt{u}) dx $

Integrate, and forget $ + C $ because it's a definite integral:

$ \left[\frac{2}{3}u^{3/2}\right]_0^2 $

We don't need to substitute the function back in, because we already evaluated it at the integrals. Substitute those instead

Evaluate:

$ \left[\frac{2}{3}2^{3/2}\right] - \left[\frac{2}{3}0^{3/2}\right] $

#### Method 1

Alternatively, instead of keeping u, substitute the original function back in

$ \int_{-1}^1 (\frac{du}{dx} \sqrt{u}) dx $

Integrate, and forget $ + C $ because it's a definite integral:

$ \left[\frac{2}{3}u^{3/2}\right]_{-1}^2 $

Substitute:

$ \left[\frac{2}{3}(x^3 + 1)^{3/2}\right]_{-1}^2 $

$ \left[\frac{2}{3}(2^3 + 1)^{3/2}\right] - 0 $

Both methods are equivalent

```octave
(2/3)*2^(3/2) - 0
```

### Integration by parts

This is the reversal of the product rule.

Given a function in the form:

$ f'(x) \cdot g(x) + f(x) \cdot g'(x) $

Then we might suppose this is a derivative obtained using the product rule.

Write down both sides of the supposed result.

$ \frac{d}{dx}[f(x) \cdot g(x) ] = f'(x) \cdot g(x) + f(x) \cdot g'(x) $

But we want to find the original function, so let's integrate both sides:

$ \int(\frac{d}{dx}[f(x) \cdot g(x) ]) dx = \int(f'(x) \cdot g(x) + f(x) \cdot g'(x))dx $

The integral of a derivative cancels on the left, so:

$ f(x) \cdot g(x) = \int(f'(x) \cdot g(x) + f(x) \cdot g'(x))dx $

Also, we know that integration is a linear operator... we can split the integral into parts.

$ f(x) \cdot g(x) = \int(f'(x) \cdot g(x))dx + \int(f(x) \cdot g'(x))dx $

So far, we've done the "parts" part of this method. Let's rearrange to get one integral by itself:

$ f(x) \cdot g(x) - \int(f'(x) \cdot g(x))dx = \int(f(x) \cdot g'(x))dx $

The above derivation is what you need to remember.

Or more concisely:

$ \int(u \cdot v')dx = u \cdot v - \int(u' \cdot v)dx $

Eg:

$ y = x \cos x $

Assume:

$ u = x $

$ v' = \cos x $

Note: It will be easier to do this if $ u $ has a simple derivative. Otherwise this method might not be useful.

Implies:

$ u' = 1 $

$ v = \sin x $

$ \int(x \cdot \cos x)dx = x \cdot \sin x - \int(1 \cdot \sin x)dx $

$ \int(x \cdot \cos x)dx = x \cdot \sin x - \int(\sin x)dx $

$ \int(x \cdot \cos x)dx = x \cdot \sin x - (-\cos x) $

Don't forget $ + C $

$ \int(x \cdot \cos x)dx = x \cdot \sin x + \cos x + C $


### Reduction formulae

Every time you integrate using Integration By Parts, you lower the power of the integrand. You can integrate several times to get something workable.

There are tables of these and you can look them up when you have something that is hard.

### Integrals of Rational Functions

Before integrating, use `Partial Fraction Decomposition` to reduce a complex fraction to two simpler ones. It should be simpler to integrate those two parts.
