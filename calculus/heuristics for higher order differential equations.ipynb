### Heuristic approaches to solving higher order ordinary differential equations

I think the mantra is that `linear is easy, non-linear is hard`. So it seems like the aim of these heuristic methods is to break a non-linear problem into a combination of linear problems.

### Homogoenous second order differntial equation

Given a problem in the form:

$ ay'' + by' + cy = 0 $

Then we can suppose that:

$ y = e^{mx} $

That implies that:

$ y' = me^{mx}, y'' = m^2e^{mx} $

Then:

$ a \cdot m^2e^{mx} + b \cdot me^{mx} + c \cdot e^{mx} = 0 $

$ e^{mx}(am^2 + bm + c) = 0 $

This is useful because now we have a quadratic function to solve:

$ am^2 + bm + c = 0 $

For any polynomial of degree 2, we know that there are two roots.

You could solve for $ m $, to find the roots for the polynomial. Those might be the same (one x intercept), different (two x-intercepts) or complex (for a polynomial with no x-intercepts).

So, if the solved values of $ m $ are $ r_1, r_2 $ then the polynomial is:

$ (m-r_1)(m-r_2) = am^2 + bm + c = 0 $

But recall that we factored out the supposed $ y = e^{mx} $. Then:

$ e^{mx}(m-r_1)(m-r_2) = 0 $

Then we have a solution to find y, but substituting the roots of the polynomial for $ m $:

$ y = C_1 e^{r_1 x} + C_2 e^{r_2 x} $

This is the general solution where $ C_1, C_2 $ are some unknown constants. To find a specific solution you would need to define the starting value of $y, x$ and then solve for $C$

I think you could use this for any degree ODE, but then you would potentially have to find complex roots.

#### Example

$ y'' - 4y' + 3y = 0 $

Then suppose:

$ y = e^{mx} $

That implies:

$ y' = me^{mx} $

$ y'' = m^2e^{mx} $

Then:

$ m^2e^{mx} - 4 \cdot me^{mx} + 3 \cdot e^{mx} = 0 $ 

$  e^{mx}(m^2 - 4m + 3) = 0 $

Then this is a quadratic function to solve. We don't even need to complete the square.

$ (m - 1)(m - 3) = m^2 - 4m + 3 = 0 $

$ e^{mx}(m - 1)(m - 3) = 0 $

Then the general solution to the linear equation is:

$ y = C_1e^x + C_2e^{3x} $

### Inhomogoenous second order differntial equation

Given a problem in the form:

$ ay'' + by' + cy = f(x) $

$ f $ is sometimes called the forcing term

To solve, let's define the following. Let:

$ y_h $ be the homogenous solution, where $ f(x) = 0 $

$ y_p $ be the ____ solution where we substitute $ f(x) $ for some value

Then: $ y(x) = y_h(x) + y_p(x) $

#### Forcing term substitutions

| f(x)                   | substitute             |
| ---------------------- | ---------------------- |
| polynomial of degree n | polynomial of degree n | 
| e^{rx}                 | Ae^{rx}                |
| \sin(kx)               | A\sin(kx) + B\cos(kx)  |
| \cos(kx)               | A\sin(kx) + B\cos(kx)  |
