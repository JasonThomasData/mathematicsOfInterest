### Newton's Method

This one is useful for finding a root of a function. Not all functions have roots that are easy to obtain from using textbook approaches, like completing the square to find a quadratic's roots.

Let:

$ x $ : be some independent variable

$ f(x) = x^e - x^2 - 100e $ : be some function smooth and continuous for which we want to find a root

$ x_{n+1} = x_n - \frac{f(x)}{f'(x)} $ : be an algorithm to generate a list of points in x that are (hopefully) closer and closer to the root.


```octave
X_ticks = 100;
X_min = 0;
X_max = 15;

X = linspace(X_min,X_max,X_ticks);

function y = f(x)
    y = x.^e + x.^2 - 100*e;
end

Y = f(X);

plot(X, Y);
hold on;
line([X_min,X_max],[0,0])
axis([0,15])
```

We should start with some x that we think should be close to the root

```octave
x_1 = 3;
y_1 = f(x_1);

plot(X, Y);
hold on;
line([X_min,X_max],[0,0])
line([x_1,x_1],[0,f(x_1)])
axis([0,15])
```

We also want to linearise the function at that first estimate.

$ f'(x) = ex^{e-1} + 2x $

Linearise at point a:

$ L(x) = f(a) + f'(a)(x - a) $

```octave
function y = f_prime(x)
    y = e*x^(e-1) + 2*x;
end

L = f(x_1) + f_prime(x_1)*(X - x_1);

plot(X, Y);
hold on;
line([X_min,X_max],[0,0])
line([x_1,x_1],[0,f(x_1)])
line(X,L)
axis([0,15])
```

To use Newton's Method to find the root, we merely have to rearrange the linearistion function

$ L(x) \approx f(a) + f'(a)(x-a) $

$ -f(a) = f'(a)(x-a) $

$ \frac{-f(a)}{f'(a)} = x - a $

$ x = a - \frac{f(a)}{f'(a)} $

$ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} $

Finding the next $ x $ will yield successevily better root approximations.

```octave
### Using the x_2

x_2 = x_1 - f(x_1)/f_prime(x_1);

L = f(x_2) + f_prime(x_2)*(X - x_2);

plot(X, Y);
hold on;
line([X_min,X_max],[0,0])
line([x_2,x_2],[0,f(x_2)])
line(X,L)
axis([0,15])
```

```octave
### Using the x_3

x_3 = x_2 - f(x_2)/f_prime(x_2);

L = f(x_3) + f_prime(x_3)*(X - x_3);

plot(X, Y);
hold on;
line([X_min,X_max],[0,0])
line([x_3,x_3],[0,f(x_3)])
line(X,L)
axis([0,15])
```

The root that this $ L(x) $ passes on $ x $ is getting very close. You would need to have a function to know when a root is close enough.
