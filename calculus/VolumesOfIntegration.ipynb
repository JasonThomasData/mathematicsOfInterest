### Volumes of Integration

If integration finds the area under a curve, then we can use that to find volume, also.

Consider a cone, in three dimensions:

```octave
[x, y, z] = cylinder(10:-1:0, 50);

#resize
x = x*.05;
y = y*.05;

surf (x, y, z);
xlabel("X");
ylabel("Y");
zlabel("Z");
xlim([-1,1]);
ylim([-1,1]);
```

If you imagine that the cone is made of an infinite series of circles stacked with their centres on the $ z $ axis then the size of each circle depends of $ z $.

Let's make $ x $ the independent variable, rotated because of convention.

```octave
# This is probably less correct than a linear transformation, but simple to do this instead

memory = x;
x = z;
z = memory;

surf (x, y, z);
xlabel("X");
ylabel("Y");
zlabel("Z");
ylim([-1,1]);
zlim([-1,1]);
```

Any circlular cross section of this cone has an area that depends on x. This is analogous to $ \pi r^2 $, where $ r $ is the unknown.

The vertical axis could be $ z $ or $ y $ in this case. We will use $ z $ to maintain the perspective.

```octave
x = [0 1; 0 1];
z = [0.5 0; 0 0];
y = [0 0; 0 0];

surf(x,y,z);
hold on
line([0,0,1], [0,0,0], [0.5,0,0])
ylim([-1,1]);
zlim([-1,1]);
xlabel("X");
ylabel("Y");
zlabel("Z");
```

Since the cone was conceptually a series of circles stacked together, then this triangle is a series of radiuses stacked together. If we remember the formula for a circle's area is $ \pi r^2 $, and since the triangle is defined as a series of those radiuses, then it's clear that the volume of the cone is:

$ V = \pi \int_0^1 T^2 dx $

Where $ T = -.5x + 1 $ is a function to define the triangle's longest side.

Notice that the ^2 is applied to the expression that is integrated, before it is integrated.

If we remember what integration represents (think of the Fundamental Theorem of Calculus, the second part) then an integral is the summation of an infinite number of rectangles with their width approaching 0.

### Example of wok

Consider a wok that's based on a sphere of radius 16cm. We think that a depth of 9cm is suffcient to hold 3 litres of water. Is this correct?

Let's first define a circle of radius 16:

$ 16^2 = (x - 16)^2 + y^2 $

But we wish the entire circle's x to be in the positive region, for simplicity:

The upper part is:

$ y^2 = 16^2 - (x - 16)^2 $

$ y = \sqrt{(16^2 - (x - 16)^2)} $

Below, we consider the integral that's shaded

```octave
function y = f(x)
    y = sqrt(16^2 - (x.-16).^2);
end

X = linspace(0,32);
Y = f(X);

plot(X, Y);
hold on
xlim([-4, 36]);
ylim([-20, 20]);
axis("square");
line([-4,36],[0,0]);
line([0,0],[-20,20]);

areaX = linspace(0,9);
areaY = f(areaX);
area(areaX, areaY)
```

Given the area of a circle:

$ \pi r^2 $

And since the shaded area is a stack of radiuses, then

$ V = \pi \int_0^9 \left(\sqrt{16^2 - (x-16)^2} \right)^2 dx $

$ V = \pi \int_0^9 \left(16^2 - (x-16)^2\right) dx $

$ V = \pi \int_0^9 \left(256 - (x^2 - 32x + 256)\right) dx $

$ V = \pi \int_0^9 \left(256 - x^2 + 32x - 256\right) dx $

$ V = \pi \int_0^9 \left(-x^2 + 32x\right) dx $

$ V = \pi \left[-\frac{1}{3}x^3 + 16x^2\right]_0^9 $

$ V = \pi \left[-\frac{1}{3}(9)^3 + 16(9)^2\right] - \pi \left[-\frac{1}{3}(0)^3 + 16(0)^2\right] $
