## Diagnosing regression

To be sure that regression is a good modelling technique for a given problem, we should check these assumptions are satisfied. These all concern the residuals:

- Correct functional form: do the independent and dependent variables have a linear relationship? (residuals vs fitted)
    - You might need to transform the data to be in a linear form
- Residuals are normally distributed
- Constant variance of residuals
- Residuals are statistically independent

At the bottom is the same data with outliers removed.

```r
# Note, the regressionExampleData has a few problems - outliers
data = read.csv("regressionExampleData.csv")
# Note, the independent variable always is on the right
model = lm(data$Exam~data$Assign)
```

#### Correct functional form (linear relationship)

A flat, straight and horizontal line at zero here would be perfect. Patterns, bumps etc suggest problems

```r
plot(model, 1)
```

### Residuals are normally distributed

Data should fall approximately on the line. Also perform some normality tests.

```r
plot(model,2)
```

```r
source("../data_science_helpers/R_helpers.R")

plot_distribution(data = model$residuals,
                  label = "Residuals",
                  distribution_is_normal = is_normal_distribution(data = model$residuals))
```

### Constant variance

It’s good if you see a horizontal line with equally spread points. If there isn't a constant variance then consider a transform of the output, perhaps a log transform.

The measure of this is often refered to as $ \sigma^2 $, to denote variance. 

```r
plot(model, 3)
```

### Statistical independence of residuals

Cook's distance indicates the influence of a value. The chart assigns labels to the top three influential values.

The Residuals vs Leverage plot can help us to find influential observations. Outlying values are generally located at the upper right corner or at the lower right corner.

```r
plot(model, 4)
plot(model, 5)
```
