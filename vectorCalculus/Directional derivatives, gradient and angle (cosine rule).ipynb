### Directional derivative

If you have a surface defined in three dimenions then you can find the rate of change of the surface's height as you move along a straight line in the $x,y$ plane.

The straight line in the $x,y$ plane would not have any z component.

Given the vector equation definition of a line:

$ (a, b) + t(\vec{u_1}, \vec{u_2}) $

Where:

$ (a, b) $ is some point in space with respect to the origin.

$ \vec{u} = u_1 i + u_2 j $ is the unit vector

$ t $ is a scaling factor

Then we can define the cartesian equation:

$ x = a + t\vec{u_1} $

$ y = b + t\vec{u_2} $

Also, let:

$ z(x,y) $ be some surface. The point of this is to find the rate of change on the surface as the direction changes.

The derivative below uses the chain rule where $ x, y $ are each functions of $ t $.

$ \frac{dz}{dt} = \frac{\partial z}{\partial x} \frac{dx}{dt} + \frac{\partial z}{\partial y} \frac{dy}{dt} $

Note the derivative of $ x, y $ are each full derivatives because they both have just one variable. But, we already know those full derivatives. They are the respective components of:

$ \vec{u} = u_1 i + u_2 j $

$ \frac{dz}{dt} = \frac{\partial z}{\partial x} u_1 + \frac{\partial z}{\partial y} u_2 $

$ \frac{dz}{dt} = \left(\frac{\partial z}{\partial x}i + \frac{\partial z}{\partial y}j\right) \cdot \left(u_1 i + u_2 j\right) $

Let this be the gradient:

$ \nabla z = \frac{\partial z}{\partial x}i + \frac{\partial z}{\partial y}j $

Let $ D_u z $ be the directional derivative.

$ D_u z = \nabla z_{(a,b)}\cdot \vec{u} $

Since this is a dot product, then it will yield a scalar value. If the scalar is positive then the surface is increasing in that direction, and if the scalar is negative then the surface is decreasing.

### Example

$ z(x,y) = 5 - (x^2 + y^2) $

We can regard $ 5 $ as a constant for the $ z $ axis, and $ z $ is a dependent variable.

$ \vec{u} = \frac{1}{\sqrt{2}}(i + j) $

$ a = (1, 1) $ 

The vector equation of the direction line is:

$ (1, 1) + t \left(\frac{1}{\sqrt {2}},\frac{1}{\sqrt {2}}\right) $

Actually I think the gradient is technically 3 partials, but since the partial of $ z $ axis is 0, then we don't need that.

The cartesian equation is:

$ x = 1 + t\left(\frac{1}{\sqrt{2}}\right)$

$ y = 1 + t\left(\frac{1}{\sqrt{2}}\right)$

```octave
a = [1 1];
u = [1/sqrt(2), 1/sqrt(2)];

# This is the surface
surfX = linspace( -2, 2 );
surfY = linspace( -2, 2 );
surfZ = 5 - (surfX.^2 + transpose(surfY.^2));

mesh(surfX,surfY,surfZ);
hold on;
axis([-2.1 2.1 -2.1 2.1 -2 5]);

# This is the direction
# it is offset by z=-2 for ease of viewing

n = 100;
t = linspace(0,1,n);

dirX = a(1) + t .* u(1);
dirY = a(2) + t .* u(2);
dirZ = -2*ones(n);

line(dirX, dirY, dirZ, "linewidth", 3);

view(-155, 24);

# This is the line on the surface that we want to differentiate:

lineZ = 5 - (dirX.^2 + dirY.^2);
line(dirX,dirY,lineZ, "linewidth", 6);
```

$ \frac{\partial z}{\partial x} = -2x $

$ \frac{\partial z}{\partial y} = -2y $

$ D_u z(1,1) = \left(-2xi + -2yj\right) \cdot \left(\frac{1}{\sqrt{2}}i + \frac{1}{\sqrt{2}}j\right) $ 

$ D_u z(1,1) = \left(\frac{-2}{\sqrt{2}} + \frac{-2}{\sqrt{2}}\right) $ 

$ D_u z(1,1) = \frac{-4}{\sqrt{2}} $ 

But that's the gradient going downslope. If we change the unit vector's sign:

$ -\vec{u} = -(\frac{1}{\sqrt{2}}i + \frac{1}{\sqrt{2}}j) $

Then:

$ D_u z(1,1) = \left(-2xi + -2yj\right) \cdot -\left(\frac{1}{\sqrt{2}}i + \frac{1}{\sqrt{2}}j\right) $ 

$ D_u z(1,1) = \frac{2}{\sqrt{2}}i + \frac{2}{\sqrt{2}}j $ 

$ D_u z(1,1) = \frac{4}{\sqrt{2}} $ 

It appears that the rate of change at $ (1, 1) $ is $ \frac{4}{\sqrt{2}} $, but that depends on the unit vector's direction.

### Angle of directional derivative (Cosine Rule)

Since the directional derivative:

$ D_u z(a, b) = \nabla z \cdot \vec{u} $

Is a dot product, and since we have this important result thanks to ths Cosine Rule:

$ \vec{a} \cdot \vec{b} = |\vec{a}| |\vec{b}| \cos \theta $

Then this implies:

$ \nabla z_{(a,b)} \cdot \vec{u} = |\nabla z| |\vec{u}| \cos \theta $

By definition, $ |\vec{u}| = 1 $. Then:

$ \nabla z_{(a,b)} \cdot \vec{u} = |\nabla z| \cos \theta $

The gradient $ \nabla z_{(a,b)} $ doesn't change. The only thing that does change here is $ \theta $.

The largest value that $ \cos $ will return is $ \cos 0 = 1 $.

Then the angle, $ \theta = 0 $, is the angle of the gradient $ \nabla z(a,b) $ with respect to the unit vector $ \vec{u} $ that maximises the gradient.

An angle of $ 0 $ is fancy talk for two parallel lines.

In summary, the directional derivative is maximal when $ \nabla z(a,b) $ and $ \vec{u} $ are parallel.

By implication the least value of $ \cos \theta $ will occur when $ \theta = \pi $, and this is in the other direction. It is parallel but in the opposite direction.

Conceptually this should be all we require to define a gradient descent algorithm, or gradient ascent if you like.
