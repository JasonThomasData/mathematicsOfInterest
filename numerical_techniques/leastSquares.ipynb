### Least Squares 

This is based on the wk2 content in the Numerical Techniques course.

```octave
x = -3:1:3;
y = [-11,-14,-10,5,6,17,14];

plot(x,y,"ro");
```

Let $ n $ be the degree function to fit to the data, and $ k \in n $





$ s_k = \sum_\limits{i=0}^m x_i^k $

$ b_k = \sum_\limits{i=0}^m x_i^k y_i $

```octave
x = [0.0, 1, 2.0, 2.5, 3.0, 4.0, 4.5];
y = [2.9, 3.7, 4.1, 4.8, 5.4, 5.8, 6.6];

n = 3;
a = leastSquares(x, y, n);

fitted = 0;
for i=1:(n+1)
    fitted = fitted + a(i)*x.^(i-1);
end

disp("Coefficients:");
disp(a);

plot(x, y, "ro");
hold on;
plot(x, fitted);
title(sprintf("Plot of y~x with regression of degree=%d", n));
xlabel("x");
ylabel("y");
```
