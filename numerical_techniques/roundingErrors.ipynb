### Errors in Guassian Elimination due to rounding

$ \frac{1}{8}x_1 + \frac{1}{7}x_2 = 1.7 $




```octave
format long;

# Long numbers

A = [1/8 1/7;
     1/9 1/8]
     
b = [17/10;
     3/2]
     
sol = A\b

```

```octave
function y = roundToDp(A, dp)
    rounding = 10^dp;
    y = round(A * rounding)/rounding;
end

trials=8
err_x1 = zeros(1,trials);
err_x2 = zeros(1,trials);
for i=1:trials;
    A_rounded = roundToDp(A, i);
    sol_rounded = A_rounded\b;
    err_x1(i) = abs(sol(1) - sol_rounded(1));
    err_x2(i) = abs(sol(2) - sol_rounded(2));
end

plot(1:trials, err_x1);
hold on;
plot(1:trials, err_x2);
title("Error as a function of decimal place rounding");
xlabel("Rounding to place");
ylabel("Error");
legend("err x1", "err x2");
```
