```r
# NOTE: I got confused and didn't get there.
# Below is the correct answer from the sheet with my own annotations.

# Firstly, note that 16 letters were sent out, 6 people subscribed.

# Size of prior, and to be the size of simulated results. Not the size of observed data
n_draw <- 10000

# This is P(A)
# Assume uniform prior
prior <- runif(n_draw, 0, 1)
hist(prior)
```

```r
# For each point in the prior, generate n=1 number from the possible numbers 0 to 16 (sample size)
gen_model <- function(rate) {
  subscribers = rbinom(n = 1, size = 16, prob = rate)
  (subscribers)
}
subscribers <- rep(NA, n_draw)
for(i in 1:n_draw) {
  subscribers[i] <- gen_model(prior[i])
}

# For each generated (simulated) data point, if it doesn't match the observed data, then remove the associated prior probability
observed_data = 6
posterior <- prior[subscribers == observed_data]

# Checking that there enough samples left
length(posterior)

hist(posterior, xlim = c(0, 1))
```

### Interpretation

The video - https://www.youtube.com/watch?v=3OJEae7Qb_o

![pic of video](Screenshot_2024-01-17_15-11-44.png)

Then explains the following:

$ P(A) $ is the probability of drawing a value $ A $ from the prior distribution.

$ P(B|A) $ is the probability that the generative model generates $ B $ given we use $ A $ from the prior.

$ P(A|B) $ is the posterior (distribution), with the prior distribution values removed if they don't help meet the criteria.

And that:

$ P(A | B) \propto P(B | A) P(A) $
