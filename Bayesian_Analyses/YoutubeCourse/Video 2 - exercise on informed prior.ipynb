### Video 2 - exercise on informed prior

From video - https://www.youtube.com/watch?v=mAUwjSo5TJE

Rather than using a uniform prior, which is an uninformed prior, let's use an informed prior.

For example, for a particular event, probability of success is observed to be $ \approx 0.2 $

```r
n_draw <- 10000

prior <- rbinom(n_draw,100,0.2)/100
hist(prior)
```

```r
successes = 6
sample_size = 16

gen_model <- function(rate) {
  subscribers = rbinom(n = 1, size = sample_size, prob = rate)
  (subscribers)
}
subscribers <- rep(NA, n_draw)
for(i in 1:n_draw) {
  subscribers[i] <- gen_model(prior[i])
}

posterior <- prior[subscribers == successes]

length(posterior)

hist(posterior, xlim = c(0, 1))
```

### Interpretation

When you use an informative prior, this will affect the result. But if you use more data then you can dilute the effect of the prior.

```r
# we need a larger prior
n_draw <- 5000000
prior <- rbinom(n_draw,100,0.2)/100

# to increase the data, otherwise the posterior would be more narrow
successes = 30
sample_size = 80

gen_model <- function(rate) {
  subscribers = rbinom(n = 1, size = sample_size, prob = rate)
  (subscribers)
}
subscribers <- rep(NA, n_draw)
for(i in 1:n_draw) {
  subscribers[i] <- gen_model(prior[i])
}

posterior <- prior[subscribers == successes]

length(posterior)

hist(posterior, xlim = c(0, 1))
```
