### Definining the prior

If you have a parameter of interest that is continuous and has a range of 0 to 1, then a Beta distribution could be appropriate for the prior. You could also use a uniform distribution but that's a non-informative prior, so with some expert information available to us then we can do a lot better.

We should specify:
- Expert information - mean, mode. These are location measures for defining a place on the distribution
- Degree of belief - variance, standard deviation, concentration. These are how spread out the shape is

Let:

$ \theta $ be the parameter of interest
$ Beta(a, b) $ be the distribution

Then:

$ \theta \approx \beta(a,b) $

To see what $a,b$ do, check out Wikipedia: https://en.wikipedia.org/wiki/File:PDF_of_the_Beta_distribution.gif

#### Definitions of Beta's location and concetration in terms of a,b

Mean (location):

$ \mu = \frac{a}{a+b} $

Mode (location):

$ \omega = \frac{a-1}{a+b-2} $

Concentration:

$ \kappa = a + b $

Variance: I have not learned how, but it is possible.


#### Defining a,b using mean

$ \mu = \frac{a}{\kappa} $

$ a = \mu \kappa $

$ -a = -\mu \kappa $

$ \kappa -a = \kappa - \mu \kappa $

$ b = (1 - \mu) \kappa $

#### Defining a,b using mode

$ \omega = \frac{a-1}{\kappa-2} $

$ \omega(\kappa-2) = a-1 $

$ a = \omega(\kappa-2) + 1 $

$ a-\kappa = \omega(\kappa-2) + 1 - \kappa $

$ -b = \omega(\kappa-2) + 1 - \kappa $

$ b = -\omega(\kappa-2) - 1 + \kappa $

#### Using variance

I have not learned how to derive these but will at least write them down for reference:

$ a = \mu \left( \frac{\mu (1-\mu)}{\sigma^2} - 1\right) $

$ b = (1-\mu) \left( \frac{\mu (1-\mu)}{\sigma^2} - 1\right) $

### Finding the posterior

Remember Bayes Theorem:

$ \textrm{posterior} = p(\theta | z) = \frac{\textrm{(likelihood)} x \textrm{(prior)}}{\textrm{constant}} $

Likelihood PDF:

$ \textrm{likelihood} = \theta^z (1-\theta)^{(n-z)} $

Prior PDF:

$ \textrm{prior} = \frac{\theta^{(a-1)} (1-\theta)^{(b-1)}}{B(a,b)} $

Posterior PDF:

$ p(\theta | z) = \frac{\theta^z (1-\theta)^{(n-z)} * \frac{\theta^{(a-1)} (1-\theta)^{(b-1)}}{B(a,b)}}{p(z)} $

$ p(\theta | z) = \frac{\theta^z (1-\theta)^{(n-z)} * \theta^{(a-1)} (1-\theta)^{(b-1)}}{p(z)B(a,b)} $

$ p(\theta | z) = \frac{\theta^{(z+a)-1} (1-\theta)^{(n-z+b)-1}}{p(z)B(a,b)} $

This final equation looks a lot like the Beta Distribution's PDF, which is:

$ \frac{x^{a-1}(1-x)^{b-1}}{B(a,b)} $

For reasons I don't understand yet, let's adopt these as the parameters from the Posterior PDF as the parameters of the Beta PDF in the numerator.

$ p(\theta | z) = \frac{1}{B(z+a, n-z+b)}\theta^{(z+a)-1} (1-\theta)^{(n-z+b)-1} $

That is the final form of the posterior.

And:

$ p(\theta | z) \propto \theta^{(z+a)-1} (1-\theta)^{(n-z+b)-1} $

Then this is the proportionality constant:

$ \frac{1}{B(z+a, n-z+b)} $

It should be clear that we have redefined $ a, b $. 

Now we can update $ \mu $ (mean), $ \omega $ (mode) and $ \kappa $ (concentration) to generate the posterior distribution.

Then let's rewrite them:

$ \mu = \frac{z+a}{z+a+n-z+b} = \frac{z+a}{a+n+b}$

$ \omega = \frac{z+a-1}{z+a+n-z+b-2} = \frac{z+a-1}{a+n+b-2} $

$ \kappa = z+a+n-z+b = a+n+b $

You can also define the posterior distribution's posterior this way.
