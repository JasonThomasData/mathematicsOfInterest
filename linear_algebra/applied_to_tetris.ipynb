### Linear algebra for controlling the rotation of tetris objects

Matrices are augmented from column vectors

```octave
STANDARD_BASIS = [1 0; 0 1]
```

```octave
# Linear transformations, aka new bases

CLOCKWISE = [0 1; -1 0]
ANTI_CLOCKWISE = [0 -1; 1 0]
```

```octave
P_1 = [1 0 0 -1; 0 0 1 0]
```

```octave
function drawPiece(piece)
    axis("square");
    axis([-2 2 -2 2]);
    hold on;
    plot(piece(1,:), piece(2,:), "or")
end
```

```octave
set(0, "defaultlinemarkersize", 90);
drawPiece(P_1)
```

```octave
P_1 = ANTI_CLOCKWISE*P_1
drawPiece(P_1)
```

```octave
P_1 = CLOCKWISE*P_1
drawPiece(P_1)
```

```octave
P_1 = CLOCKWISE*P_1
drawPiece(P_1)
```

```octave
P_2 = [1 1 0 -1; 1 0 0 0]
```

```octave
drawPiece(P_2)
```

```octave
P_2 = CLOCKWISE*P_2
drawPiece(P_2)
```

```octave
P_2 = CLOCKWISE*P_2
drawPiece(P_2)
```

```octave
P_3 = [1 0 0 -1; 1 1 0 0]
drawPiece(P_3)
```

```octave
P_3 = ANTI_CLOCKWISE*P_3
drawPiece(P_3)
```

```octave
P_3 = ANTI_CLOCKWISE*P_3
drawPiece(P_3)
```
