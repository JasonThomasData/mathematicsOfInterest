### Linear algebra for controlling the rotation of tetris objects

Matrices are augmented from column vectors

```octave
STANDARD_BASIS = [1 0; 0 1]
```

```octave
P_1 = [1 0 0 -1; 0 0 1 0]
```

```octave
set(0, "defaultlinemarkersize", 90);
axis("square");

axis([-2 2 -2 2]);
hold on;
plot(P_1(1,:), P_1(2,:), "or")
```

```octave
# Linear transformations, aka new bases

CLOCKWISE = [0 1; -1 0]
ANTI_CLOCKWISE = [0 -1; 1 0]
```

```octave
axis("square");
P_1 = ANTI_CLOCKWISE*P_1
axis([-2 2 -2 2]);
hold on;
plot(P_1(1,:), P_1(2,:), "or")
```

```octave
axis("square");
P_1 = CLOCKWISE*P_1
axis([-2 2 -2 2]);
hold on;
plot(P_1(1,:), P_1(2,:), "or")
```

```octave
axis("square");
P_1 = CLOCKWISE*P_1
axis([-2 2 -2 2]);
hold on;
plot(P_1(1,:), P_1(2,:), "or")
```
