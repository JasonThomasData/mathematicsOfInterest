### Defining a plane

Remember this: the cross product of any two perpendicular vectors is equal to 0.

#### Using a vector normal

A vector normal is a vector that is perpendicular to a plane, or intuitively the plane's facing side.

Let's define it as $ \vec{N} = (x_n, y_n, z_n) $

Take any point in the plane, call it $ P_0 = (x_0, y_0, z_0) $. This should have values defined.

Then take another arbitrary point on the plane. This one needs to be symbols only. Call it $ P_a = (x_a, y_a, z_a) $

Then the vector between those is $ \vec{P_0 P_a} = (x_a - x_0, y_a - y_0, z_a - z_0) $. Clearly this vector must be on the plane. 

That means that the vector normal must be perpendicular to that vector:

$ \vec{P_0 P_a} \cdot \vec{N} = 0 $

$ (x_a - x_0, y_a - y_0, z_a - z_0) \cdot (x_n, y_n, z_n) $

$ x_n(x_a - x_0) + y_n(y_a - y_0) + z_n(z_a - z_0) = 0 $

$ x_n(x_a - x_0) + y_n(y_a - y_0) + z_n(z_a - z_0) = 0 $

$ x_n * x_a - x_n * x_0 + y_n * y_a - y_n * y_0 + z_n * z_a - z_n * z_0 = 0 $

$ x_n * x_a + y_n * y_a  + z_n * z_a = x_n * x_0 + y_n * y_0 + z_n * z_0 $ 

All of the values on the right are scalars and can be collapsed into a single value.

#### Example

$ \vec{N} = (4, 2, -1) $

$ \vec{P_0} = (1, 3, -2) $

$ \vec{P_a} = (x, y, z) $

$ \vec{P_0 P_a} = (x-1, y-3, z+2) $

$ (x-1, y-3, z+2) \times (4, 2, 1) = 0 $

$ 4(x-1) + 2(y-3) + 1(z+2) = 0 $

$ 4x-4 + 2y-6 + z+2 = 0 $

$ 4x + 2y + z = 8 $

#### Using three points

If we have any three points in space then we can use those to define the vector normal of a plane that contains those three points.

This is once again achieved by using the cross product since the cross product of any two vectors is perpendicular to both vectors. Since the the vectors to cross are in the plane then this will yield the vector normal.

#### Example

Suppose we have any three points:

$ A = (1, 2, 3) \\
B = (-1, 5, -1) \\
C = (3, 4, 5) $

Then you can deduce that these vectors exist on the plane also:

$ \vec{AB} = (-1 - 1, 5 - 2, -1 - 3) = (-2, 3, -4) \\
\vec{AC} = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2) $

Then the vector normal for the plane is the cross product: 

$ \vec{N} = \vec{AB} \times \vec{AC} $

We find it by finding the determinant:

$ \begin{matrix} i & j & k \\ -2 & 3 & -4 \\ 2 & 2 & 2 \end{matrix} $

$ i(3*2 - -4*2) - j(-2*2 - -4*2) + k(-2*2 - 3*2) $

$ \vec{N} = 14i - 4j -10k $

That's the vector normal. Keep one of either $ A,B,C $ and then define the plane using a point and $ \vec{N} $.
