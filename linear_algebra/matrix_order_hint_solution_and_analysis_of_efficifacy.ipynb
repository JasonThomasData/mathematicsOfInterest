### Row and column rank as a hint: none, one or infinitely many solutions

DISCLAIMER: This is an idea and may not ever get finished

It is important to recognise that for any system of equations, and then turned into an augmented matrix, the row rank will always equal the column rank. We can weaponise this information to create a visualisation that addresses this concept: is it possible to guess if a system of equations has none, one or infinitely many solutions?

The answer is yes, since the order of a matrix is somewhat useful as a predictor for this. I would like to see how reliable such an indicator is.

The approach will be: given a range n=1:3, m=1:3, then for n rows and m columns, and given a range p=-6:1:6, then for each variant of p as a coefficient in a matrix of order n\*m, then can we count the matrices that fall into the categories mentioned (none, one, infinite).



The general idea is: given a matrix augmented with a zero vector, therefore in the form that you would check to see if a matrix is linearly independent or not, what can we guess from the order of the matrix?

Generally:

$ \begin{matrix} order & outcome \\ row<column & infinite \\ row=column & one \\ row>column & none \end{matrix} $

Then, it would be interesting to see the probability of each over a given range:

$ \begin{matrix} & wide & square & tall \\ none & . & . & . \\ one & . & . & . \\ infinite & . & . & . \end{matrix} $

```octave
# 2x2
n = 2;
m = 2;
count = 0;
lower=-5;
upper=5;
step=1;
start = time;
matrix = zeros(n,m);

# a,b
# c,d

for a=lower:step:upper
    matrix(1,1)=a;
    for b=lower:step:upper
        matrix(1,2)=b;
        for c=lower:1:upper
            matrix(2,1)=c;
            for d=lower:1:upper
                matrix(2,2)=d;
                linsolve(matrix,[0;0]);
                if(rank(matrix) == 2)
                    count = count + 1;
                end
            end
        end
    end
end
finish = time;
```

This became complex quickly, even for just a 2x2 matrix with values ranging -5:5. For this to be viable at all, you'd need to restrict either the range of values to where it is with a maximum size of 3x3, or, restrict the range of values to a smaller set for a larger variety of matrices.
