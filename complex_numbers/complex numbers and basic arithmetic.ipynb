### Complex numbers

#### A definition for complex numbers

Recall that

$ i^0 = 1 $

$ i^1 = i $

And then observe the important definition:

$ i^2 = -1 $ (let this be a convenient reference for defining others)

Then by implication:

$ i^3 = i \cdot i^2 = -i $

And that is the cycle. It repeats:

$ i^4 = i^2 \cdot i^2 = 1 $

$ i^5 = i^3 \cdot i^2 = -i \cdot -1 = i $

$ i^6 = i^4 \cdot i^2 = 1 \cdot -1 = -1 $

$ i^7 = i^5 \cdot i^2 = i \cdot -1 = -i $

And on it goes with this pattern

### Arithmetic with complex numbers

Mostly you will see complex numbers expressed with a real $ a $ imaginary $ b $ parts.

$ (a + bi) $, eg: $ (4 + 2i) $

Treat all complex numbers as vectors with real and imaginary elements.

#### Addition

$ (a + bi) + (c + di) = (a+c) + (b+d)i $

Example:

$ (5 + 2i) + (-1 - 8i) $

$ (5-1)+(2-8)i $

$ (4-6i) $

$ 2(2-3i) $

Subtraction is obviously the same but is the inverse operation.

#### Multiplication

$ (a + bi)(c + di) $

$ ac + adi + cbi + bd(i)^2 $

$ ac + adi + cbi + bd(-1) $

$ ac + adi + cbi - bd $

Eg:

$ (3 - 2i)(4 + 4i) $

$ 12 + 12i - 8i - 8i^2 $

$ 12 + 4i + 8 $

$ (20 + 4i) $

$ 4(5 + i) $

#### Division

$ \frac{(a + bi)}{(c + di)} $

To do division, first find the complex conjugate of the denominator.

It flips sign for the imaginary part on the imaginary plane. Then multiply it to the denominator and numerator.

$ \frac{(a + bi)}{(c + di)} \frac{(c - di)}{(c - di)} $

Then follow the normal rules for multiplication

Eg:

$ \frac{(3 - 2i)}{(2 + 4i)} $

$ \frac{(3 - 2i)}{(2 + 4i)} \frac{(2 - 4i)}{(2 - 4i)} $

$ \frac{6-12i-4i+8i^2}{4-8i+8i-16i^2} $

$ \frac{6-16i-8}{4+16} $

$ \frac{-2-16i}{20} $

$ \left(-0.1 - 0.8i\right) $

$ \frac{-1}{10} \left(1 + 8i\right) $
