### Test space

This is a space we use to test connectedness.

Let the set:

$ \underline 2 = \{0, 1\} $

And let's define $ (\underline 2, \mathcal{T}) $ as the set with the discrete topology.

We test it using this theorem:

A topological space $ (Y, \mathcal{U}) $ is connected iff every continuous function $ f: X \to \underline 2 $ is constant. This means every element to $ 0 $ or every element to $ 1 $.

This is continuous:

![constant](Screenshot_2023-04-12_16-23-07.png)


The notion above says that if you define a way to say a point in $ X $ maps to $ 0 $, then everything around it that's arbitrarily close to that point also goes to $ 0 $. If something is not arbitrarily close then it must be in a disjoint set, therefore not connected.

To use the test space, we need to first know that the function is continuous.

If it is, then think about the set $ P \left(\underline 2 \right) = \{ \emptyset, \{0\}, \{1\} \{0, 1\} \} $ .

Now think about the pre-image of these.

$ f^{-1}(\emptyset) = \emptyset $

$ f^{-1}(\{0\}) = A $

$ f^{-1}(\{1\}) = B $

$ f^{-1}(\{0,1\}) = f^{-1}\left(\underline 2 \right) = X $

If the function maps to either elements of the test space then $ X $ must be disjoint, but if it maps to the test space itself then $ X $ must be connected.

If you read the statements from earlier that are equivalent (6 of them) then if the space is connected then either A or B must be the empty set and then the other must map to the entire test space.

### Disconnectedness

The antithesis of connectedness is disconnectedness. This concept, aka "totally disconnected" is not the same concept of disconnected mentioned earlier in the notebook.

A subset $ A $ of a topological space $ (X, \mathcal{T}) $ is totally disconnected if every connected component is a singleton set.

#### Example - discrete topology

If $ X $ has the discrete topology then every element of $ X $ is a singleton set. Therefore the discrete topology is totally disconnected.

#### Proof that the rational numbers are totally disconnected

The rational numbers $ \mathbb{Q} \subseteq \mathbb{R} $ are totally disconnected.

To see this, fix $ x < y \in Q $. It doesn't matter how small the difference is, you can always fit an irrational number between them:

$ c \in \mathbb{R} $ \ $ \mathbb{Q} : x < c < y $

Then take the open intervals including $ c $:

$ A = (-\infty, c) \cap \mathbb{Q} $

$ B = (c, \infty) \cap \mathbb{Q} $

Then $ A, B $ satisfy the requirements for being disconnected:
- $ A \cap B = \emptyset $
- $ A \cap B = \mathbb{Q} $
- Both are open
- $ A, B \ne \emptyset, \mathbb{Q} $

THIS SOMEHOW IMPLIES THAT ALL CONNECTED COMPONENTS ARE DISCONNECTED, SO THE SET IS TOTALLY DISCONNECTED

### Path-connectedness (stronger form of connectedness)

This is a stronger notion of connectedness.

The idea is that if there are two points in a space then there is a path between them that does not leave the space.

First you need to define what a "path" is. 

Given a topological space $ (X, \mathcal{T}) $

Let $ a, b \in X $ be the points

The path from $ a \to b $ is a function:

$ \gamma: [0,1] \to X $

This satisfies:

$ \gamma(0) = a $

$ \gamma(1) = b $

$ X $ is "path-connected" if there is a path between any two points $ a,b $

#### A continuous image of a path connected space is path connected

Let $ (X, \mathcal{T}) $ be a path-connected topological space.

$ f: X \to Y $ is a continuous function.

Fix $ y_1, y_2 \in f(X) $

Then $ x_1, x_2 \in X : f(x_1) = y_1, f(x_2) = y_2 $

The definition for path-connected is given by:

$ \gamma: [0,1] \to X $

$ \gamma(0) = x_1 $

$ \gamma(1) = x_2 $

But the function $ f: X \to Y $ is continuous, so:

$ f(\gamma(0)) = y_1 $

$ f(\gamma(1)) = y_2 $

Then this means the image is path-connected.

#### Every path-connected space is connected

Proving this makes sense because path-connected is a stronger term than connected, so therefore everything that's in the path connected set is in the connected set.

Fix $ a \in X $. For all $ x \in X $ there is a path:

$ \gamma: [0, 1] \to X $ that satisfies:

$ \gamma(0) = a $

$ \gamma(1) = x $

Then:

$ G_x = \gamma([0,1]) $, meaning that $ G $ is the path.

$ x \ne y $

$ (\{a\} \subseteq G_y \cap G_x) \ne \emptyset $

Each $ G_x $ is connected (to $ a $).

And that means

$ X = \bigcup_\limits{x \in X} G_x $ is connected, because the union of connected components is connected.

#### But the converse is not true!

The converse reads:

Each path-disconnected space is disconnected... this is false.
