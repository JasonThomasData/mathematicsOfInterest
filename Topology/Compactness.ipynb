### Compactness

This is analogous to finite but in the domain of Topology.

Imagine the open interval $ (0,1) \subset \mathbb{R} $

Then this function is continuous

$ (0, 1) \to (-\infty, 0) : x \to \log x $

Then this is its inverse:

$ (-\infty, 0) \to (0, 1) : x \to e^x $

Both functions are continuous and so this is a homeomorphism.

An observation we can make is this: despite $ (0,1) $ is bounded by $ 1 $, which means $ x < 1 $, $ (-\infty, 0) $ is not bounded. So we can say this is not bounded.

But this happens because the domain to start with was open. If it was closed, then it would be bounded.

So this means that open and bounded is not a topological property, but closed and bounded is a topological property.

_We can think of this notion "closed and bounded" as compactness._

#### Relation to homeomorphism

If one space is homeomorphic to another, then the spaces share the property of compactness. If one is closed and bounded then so is the other.

#### A fundamental property in Topology

Let $ (X, \mathcal{T}) $ be a topological space.

$ \{ G_\lambda : \lambda \in \Lambda\} $ be a collection of subsets, might be open and/or closed.

The collection of subsets is a "covering" (noun) if:

$ X = \bigcup_\limits{\lambda \in \Lambda} G_\lambda $

It's a "finite cover" if the indexing set $ \Lambda $ is also finite. This doesn't imply anything about each the cardinality of each individual $ G_\lambda $.

It's an "open cover" if the elements $ G_\lambda $ are open.

A "sub cover" is a subset of the $ G_\lambda $. This is:

$ \{ G_\mu : \mu \in M \} $

Where $ M \subset \Lambda $

#### Given all of these, above

We can say $ (X, \mathcal{T}) $ is compact if every open cover admits a finite subcover.

A subset $ A \subseteq X $ is compact if it's compact with respect to the subspace topology.

#### Intuition for this

Let $ X $ be compact

$ \{ G_\lambda : \lambda \in \Lambda \} $ is an open cover. Then every $ G_\lambda \in \mathcal{T} $, because each $ G_\lambda $ is open. Recall that the topology $ \mathcal{T} $ is all of the open sets in $ X $.

Then to show that $ X $ is compact we would need to show that the indexing set $ \Lambda $ is finite, such that:

$ \Lambda = \{ \lambda_1, \lambda_2, ..., \lambda_n \} $

$ X = \bigcup_\limits{i = 1}^n G_{\lambda_i} $

Above reads (G sub (lambda sub i)).

An open interval can never be covered by open subsets so it can never be compact. This doesn't count as a proof and it's just some intuition.

#### Examples

- Every finite topological space is compact
- $ \mathbb{R} $ with the Euclidean topology is not compact

In summary, compactness is not a property of a set but a property of a topological space, and it's dependent on what the topology is.
