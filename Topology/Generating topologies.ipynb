### Generating topologies

If you have some set and you have some functions that you would like to be continuous, then can we define a topology so that those functions become continuous?

The product topology has two functions. Each projects a cartesian coordinate (for example in $\mathbb{R}^2$) to its $ x $ or $ y $ component. We define it by taking the most coarse topology so that the functions are continuous.

#### Example

$ S \subseteq P(X) $ The topology on $ X $ generated by $ S $ is the coarsest (smallest) topology on $ X $ such that every element of $ S $ is open.

#### Generating a topology from an intersection of topologies

The intersection of topologies is another topology. The union of topologies isn't necessarily a topology.

Let $ \{ \mathcal{T}_\lambda : \lambda \in \Lambda \} $ be a non-empty set of topologies on $ X $. 

Then $ \mathcal{T} = \bigcap_\limits{\lambda \in \Lambda} \mathcal{T}_\lambda $

In other words, $ G \subseteq X $ is open wrt $ \mathcal{T} $ iff $ G $ is open wrt every $ \{ \mathcal{T}_\lambda : \lambda \in \Lambda \} $

This doesn't violate the third axiom because we're taking the intersection of sets of open sets, and it's not an intersection of open sets.
