### Fundamental Theorem of Algebra

Every non-constant polynomial of one variable with complex coefficients is a product of linear terms. 

This works for real coefficients too because those are complex numbers with 0 for their imaginary parts.

$ P(z) = c_1 z^1 + c_2 z^2 + c_3 z^3 + ... + c_n z^n $

$ P(z) = \sum_\limits{i=1}^n c_i z^i , c_n \ne 0, n \ge 1 $

$ P(z) = (z - \zeta_1)...(z - \zeta_n) $, where $ (z - \zeta) $ is a linear term.

One important implication is that if you are given a polynomial to any degree then you can find its roots, and the number of roots you should expect is the same degree as the polynomial.

Example:

$ x^2 + 3x + 2 = (x + 1)(x + 2) $

The polynomial above has real roots but it works also for complex roots.

#### Lemma

Let:

$ P(z) = \sum_\limits{j=1}^n c_j z^j , c_n \ne 0, n \ge 1 $

Then either $ P(0) = 0 $ or there is some complex number $ \gamma \in \mathbb{C} $ such that $ |P(\gamma)| < |P(0)| $. So I think this means that if $ 0 $ in the domain has an image other than $ 0 $, then $ 0 \le \gamma < P(0) $

#### Proof

Suppose $ P(0) $, then every term is going to be of the form: $ c_j 0^j $, and all of them will be $ 0 $. The only term that remains is the final constant that has no $ z=0 $. So therefore $ P(0) = c_0 $

Suppose $ P(0) \ne 0 $. Then we know $ c_0 \ne 0 $

Let 

$ k > 0 $ be the smallest exponent such that $ P(z) \ne c_0 $. I think this means that $ k = 1 $

$ P(z) = c_0 + c_k z^k + z^{k+1} g(z) $

$ g(z) $ is just some polynomial that is defined as having $ z^{k+1} $ factored out.

---

An example is:

$ f(z) = 1 + x^2 + 3x^3 - x^4 $

The coefficients are:

$ (c_0, c_1, c_2, c_3, c_4) = (1,0,1,3,-1) $

The smallest exponent such that the function doesn't return $ c_0 $ is $ k=2 $

So now we factor out $ x^{k+1} $ and it becomes:

$ f(z) = 1 + x^2 + x^3 (3 - x) $

And so in this example, $ g(x) = (3 - x) $
