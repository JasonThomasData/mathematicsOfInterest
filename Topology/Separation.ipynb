### Separation

Open sets give us a notion of closeness without assigning a metric to describe distance between points.

We can also use open sets to tell us when points can be separated.

Given some topological space $ (X, \mathcal{T} ) $, and $ x \ne y \in X $, then is there an open set $ G $ with $ x \in G $ but $ y \not \in G $

Then these two points are not arbitrarily close. Then this is one notion of being separate.

![separation axioms](separation.png)

#### Separation axioms to separate points

A topological space $(X, \mathcal{T})$ falls into 1 of these categories, but only one, or 0 of them (therefore not separate).

The strength of these three is given by $ < $ meaning weaker : $ T_0 < T_1 < T_2 $

And the stronger implies the weaker: $ T_2 \to T_1 \to T_0 $

- $ T_0 $ iff given $ x \ne y, G \in \mathcal{T} $ then there is either: $ x \in G, y \not \in G $ _OR_ $ x \not \in G, y \in G $

- $ T_1 $ iff given $ x \ne y, G \in \mathcal{T}, H \in \mathcal{T} $ then: $ x \in G, y \not \in G $ _AND_ $ x \not \in H, y \in H $. Obviously G, H can be switched to satisfy the axiom. $ G, H $ can have an intersection and don't necessarily need to be disjoint.

- $ T_2 $ iff in addition to the requirements for $ T_1 $, then $ G \cap H = \emptyset $, in other words the intersection is empty.

Notes: $ T_1 $ spaces sometimes known as Frechet spaces, $ T_2 $ spaces sometimes known as Hausdorff spaces. Most spaces are $ T_2 $ at least, and all metric spaces are $ T_2 $

#### Separation axioms to separate closed sets

These two are different to the ones above and there's no notion of these below implying any other.

$ T_4 \not \to T_3 $

$ T_3 \not \to T_2 $

- $ T_3 $ iff given any closed set $ F \subseteq X $, $ x \not \in F $, there are $ G, H \in \mathcal{T} $ open sets such that $ (F \subseteq G) \cap (x \in H) \cap (G \cap H = \emptyset) $

- $ T_4 $ iff given any closed sets $ F, K \subseteq X $, there are $ G, H \in \mathcal{T} $ open sets such that $ (F \subseteq G) \cap (K \subseteq H) \cap (G \cap H = \emptyset) \cap (F \cap K = \emptyset) $
