$ a \in X $

$ r > 0 $

The nice thing about an open ball is that $ \forall x \in B(a;r) $, there is a radius $ s > 0 $ where $ B(x;s) \subset B(a;r) $.

### Closed sets

A set is closed if its complement is open, and vice-versa.

A major misconception is that sets are either open or closed, but this is not true. A set can be neither open nor closed. For example subsets of a topological space can be not open or closed. They can also be open and closed... according to the lecturer, not sure how.

#### Number line intuition for closed

If $ [a,b] \subseteq \mathbb{R} $ is closed then:

$ [a,b] = \{ x \in \mathbb{R} : x > a \wedge x < b \} $

$ [a,b] = \{ x \in \mathbb{R} : \sim (x < a \vee x > b) \} $

$ [a,b] = \mathbb{R} \space \text{\\} \space \{ x < a \vee x > b \} $

$ [a,b] = \mathbb{R} \space \text{\\} \space ((-\infty, a) \cup (b, \infty)) $

Then this leads to the definition for closed:

$ (X, \mathcal{T}) $ is a topological space.

$ A \subseteq X $ is closed if $ A $ \ $ X $ is open.

#### All closed sets

$ (X, \mathcal{T}) $ is a topological space

$ F \subseteq P(X) $, be the set of all closed subsets.

Then the following are true (and in a sense backwards, compared to the axioms for a topological space.

### The closure of a set 

Let $ A \subseteq X $, sucn that $ (X, \mathbb{T}) $ is a topological space. $ A $ is any set, open or not.

The closure of $ A $ in $ X $ is $ \bar{A} $ is the smallest closed set containing $ A $. 

$ \bar{A} $ is the closure of $ A $ in $ X $

Every subset has a unique closure.

The proof: 

Let $ F \subseteq P(X) $ be all closed sets containing $ A $.

We know that $ \emptyset, X \in F $ (see the definition of a closed set). Then $ F \ne \emptyset $ because it must contain $ X $.

Let:

$ \bar{A} = \bigcap_{K \in F} K $, so in other words we intersect all of the closed sets that contain $ A $. Then $ \bar{A} $ is closed (because it's an intersection of closed sets). Then all $ K $ contain $ A $, because of the definition for $ F $.

To show $ \bar{A}$ is the smallest, if $ J $ is any other closed set that contains $ A $, then $ J \in F $, and $ \bar{A} \in J $.

#### When A is closed

$ A $ is closed iif $ A = \bar{A} $

### Density (defined using closure)

Notionally, density is where you have a set with points arbitrarily close to another set.

$ B \subseteq X $

We use the definition from before: 

We can say $ A \subseteq X $ is dense in $ B $ if $ \bar{A} = B $

##### Example

$ \mathbb{Q} $ is dense in $ \mathbb{R} $. There's one closed set that contains all of the rational numbers, and it is all $ \mathbb{R} $. But also $ \mathbb{R} \text{\\} \mathbb{Q} $ is dense in $ \mathbb{R} $

### Characterise the points in a closed set

Let $ A \subseteq (X, \mathcal{T}) $.

Then $ x \in \bar{A} $ iff every open set $ G : x \in G $ satisfies $ G \cap A \ne \emptyset $

$ \bar{A} \not \subseteq X \text{\\} G \Leftrightarrow A \cap G \ne \emptyset $

Then since $ x \in A, x \in G $, it follows that the intersection is not empty

Proof of it:

---

We wish to show:

$ \bar{A} \not \subseteq X \text{\\} G \Leftrightarrow A \cap G \ne \emptyset $

First let's show:

$ \bar{A} \not \subseteq X \text{\\} G \Rightarrow A \cap G \ne \emptyset $

---

Suppose $ x \in \bar{A} $, fix $ G \in \mathcal{T} $ such that $ x \in G $. Then $ G $ is open for it be in $\mathcal{T} $

Then $ \bar{A} \not \subseteq X \text{\\} G $, because that would mean $ x $ is one and not the other, but we defined them to contain $ x $.

$ \bar{A} \not \subseteq X \text{\\} G \Rightarrow A \not \subseteq X \text{\\} G $

$ \bar{A} \not \subseteq X \text{\\} G \Rightarrow A \cap G \ne \emptyset $

---

Next let's show:

$ A \cap G \ne \emptyset \Rightarrow \bar{A} \not \subseteq X \text{\\} G $

But we can write this as the contrapositive:

$ \sim(\bar{A} \not \subseteq X \text{\\} G) \Rightarrow \sim(A \cap G \ne \emptyset)$

$ \bar{A} \subseteq X \text{\\} G \Rightarrow A \cap G = \emptyset $

_Therefore to prove the equivalence $ P \Leftrightarrow Q $ we show $P \to Q$ and then show its converse $\sim P \to \sim Q $._

Suppose $ x \not \in \bar{A} $. Suppose there is some closed set $ K \subseteq X $ such that $ x \not \in K $, $ A \subseteq K $. This is what's meant by $ \bar{A} \subseteq X \text{\\} G $

$ X \text{\\} K $ is open since $ K $ is closed.

$ G = X \text{\\} K $

$ x \in G $, but that means $ x $ is in an open set.

But if $ x \not \in \bar{A} $ then $ x \not \in A $

Then this shows:

$ \bar{A} \subseteq X \text{\\} G \Rightarrow A \cap G = \emptyset $

### Accumulation points (point of accumulation)

An element $ x $ is a point of accumulation of a subset $ A $ if given any open set $ G $ that contains $ x $, we have that $ \{ G \cap A \} $ \ $ \{x\} \ne \emptyset $

All points of accumulation are caled the derived set $ A' $

We can think of this as a boundary that we define to form the closure of a set.

If you want to form a closure then you take a set and you add these points of accumulation.

$ \bar{A} = A \cup A' $, is the closure of a set subset $ A $.
