### Sequences, convergence, completeness

A sequence in a set $ X $ is a list of elements of $ X $ that are indexed by the natural numbers. This implies there is order.

A sequence can also be thought of as a function $ f: \mathbb{N} \to X $. It's a function from the natural numbers to the set $ X $. Again this implies order.

$ f(n) = x_n $

Then we can think of convergence here as well.

In an analysis course we could say:

_A sequence $ x_n $ converges to a limit $ a \in x $ iff $ \exists n, \forall \epsilon > 0 , \forall M > n : |x_M - a | < \epsilon $_

And we can also think of it like this:

$ a - \epsilon < x_n < a + \epsilon $

But instead of this absolute difference, we can define a metric.

_A sequence $ x_n $ converges in a metric space $ (X, \rho) $ to a limit $ a \in x $ iff $ \exists n, \forall \epsilon > 0 , \forall M > n : \rho(x_M, a) < \epsilon $_

To take a more topological bent on this notion of convergence:

$ \rho(x_n, a) < \epsilon $ iff $ n_n \in B(a; \epsilon) $, as in, an open ball.

What if instead of an open ball we say it's an open set?

_A sequence $ x_n $ converges in a topological space $ (X, \mathcal{T}) $ to $ a \in X $ iff for any open set $ G: a \in G, \forall M > n : x_M \in G $_

#### Example, convergence in a topological space

Let:

$ x_n = \frac{1}{n} $. In $ \mathbb{R} $ with the Euclidean Topology, $ \lim_\limits{n \to \infty} \left( \frac{1}{n} \right) = 0 $

But then consider the sequence in $ (0, 1] $, and give it the subspace topology.

The sequence doesn't converge in the subspace because the limit point $ 0 \not \int (0, 1] $

It doesn't converge over the reals but not in the subspace.

#### Eventually constant

Theorem: In a discrete topological space, a sequence converges iff it is eventually constant.

_The intuitive notion of eventually constant is that there is some element in the sequence such that index, all following elements are equal._

__Proof:__

Let $ (X, \mathcal{T}) $ be any set with the discrete topology.

Let $ S = \{ s_1, s_2, s_3,... s_n \} $ be a sequence, indexed by $ n \in \mathbb{N} $ 

---

1) Suppose that in the space the sequence converges.

$ \forall x \in X: {x} \in \mathcal{T} $, because we're using the finest topology.

Then one of the singleton sets in $\mathcal{T} $ is the value that the sequence converges to.

That the sequence converges implies the sequence is eventually constant.

---

2) Suppose that the sequence is eventually constant

Then the limit of the sequence, as the index $ n \to \infty $, is some element in $ X $.

The discrete topology is the finest topology so $ \{x\} \in \mathcal{T} $

Then if the limit of the sequence is in the topology, we say this converges in the topological space.

---

`You need to apply the definition of convergence more precisely. Namely, for any open neighbourhood U of the limit point x, there exists an N such that n>N implies x_n in U. Then specifically choosing U = {x} shows the sequence is eventually constant.`

#### Theorem: $ X $ converges (in a top. space) iff $ s $ extends to a continuous function

A sequence $ s: \mathbb{N} \to X $, (also written $ s(n) = x_n $), in a topological space $ X $, converges (in the space) iff $ s $ extends to a continuous function:

$ \hat s: \mathbb{N} \cup \{\infty\} \to X $

Such that $ \hat s(n) = s(n) $

Then $ \hat s $ has another point that $ s $ doesn't have. To maintain the continuity, then $ \infty $ in the codomain maps to the limit of the sequence (in the topology).

If $ s $ converges in the topological space then the point in the topology that it converges to is $ \hat s(\infty) $

$ \lim_\limits{n \to \infty} \left( s(n) \right) = \hat s(\infty) $

#### Cauchy sequences in metric spaces

Recall that a Cauchy Sequence allows us to think about convergence without defining a limit specifically.

The same notion can be applied to metric spaces.

This is the standard definition of Cauchy Sequence, but instead of absolute value, we use a metric (as in, this could be the absolute value metric but it's more abstract).

A sequence $ a_n $ in a metric space $ (X, \rho) $ is Cauchy __iff__ for all $ \epsilon > 0 $, there exists some $ N \in \mathbb{N} $ such that $ \rho(x_m - x_n) < \delta $ for all $ m,n > N $
