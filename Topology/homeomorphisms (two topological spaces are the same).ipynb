### Homeomorphism (prove two topological spaces are the same)

Let:

$ (X, \mathcal{T}), (Y, \mathcal{U}) $ be topological spaces

$ f: X \to Y $ be a continuous function

$ g: Y \to X $ also be continuous

Then that means:

$ f \cdot g = id_y $

$ g \cdot f = id_x $

This is essentially the classic example of squishing a bowl to be a plate or a cube to a sphere etc.

#### Similar to isometry

This is similar to but weaker than an isometry.

I think that if you have two metric spaces, and then each is then a topology induced from the metric, and if there is a metric-preserving function between the metrics, then we can also say there's is a homeomorphism between the induced topologies.

But being weaker, a homeomorphism between two topogical spaces doesn't imply an isometry between the metric spaces.

As an example, there is a homeomorphism from the interval $ (0, 1) $ to $ \mathbb{R} $, and we can show that a modified tan function will do the job (move the interval $(-\pi/2, \pi/2) \to (0,1) $ and the tan function will work). The inverse of the modified tan will take $ \mathbb{R} $

#### Example - a homeomorphism

There is a homeomorphism if $ [a,b) \to (a,b] $

Let there be some point $ a < c < b $

$ f(x) = c + (c - x) $

$ f(x) = 2c - x $

Then the inverse is:

$ y = 2c - x $

$ x = 2c - y $

So the function is its own inverse, and I looked it up, it's know as an involution. This is homeomorphic.

#### Not a homeomorphism

Let:

$ (X, \mathcal{T}), (Y, \mathcal{U}) $ be topological spaces.

$ S^1 = \{(x,y) \in X^2 + Y^2 = 1 \} $ be all the points $ (x,y) $ on a circle.

$ f: [0, 1) \to S^1 $ be a continuous function and is a bijection. Imagine a line mapped onto a circle. The interval must be closed on one side otherwise it wouldn't be injective and therefore not bijective.

But $ f^{-1} $ is not continuous, so this is not a homeomorphism.

Since $ f $ is given as continuous, then we know that $ f^{-1} $ is open. But $ f^{-1} $ is not continuous.

$ f{-1} : S^1 \to [0, 1) $

To show this, let $ G = [0, 1/2) $. This is an open subset of $ X $, but only because $ X $ doesn't have any negative numbers.

If $ f^{-1} $ was continuous then its preimage would be continuous. The preimage of $ f^{-1} $ is $ f $.

$ f^{-1}(f^{-1}(G)) = f(G) $

Have a look at what we've got so far. This image should ideally include the axes for the unit circle.

Notice that the $ y $ compnent is always positive and strictly larger than 0. This means that we include another point:

$ f(G) = \{(x, y) \in S^1: y > 0 \} \bigcup \{(1,0)\} $ 

That point $ (1, 0) $ is not in $ S^1 $. This means that there is no open set that contains $ (1, 0) $ that also contains $ f(G) $

Then if it's not continuous in both ways then it can't be a homeomorphism.

### An abstract homeomorphism

$ X = Y = \{a, b, c\} $

$ \mathcal{T} = \{\emptyset, \{a\}, \{a,b\}, X \} $

$ (X, \mathcal{T}) $ is a topological space.

$ \mathcal{U} = \{\emptyset, \{c\}, \{a,c\}, Y \} $

$ (Y, \mathcal{U}) $ is a topological space.

$ f: X \to Y $

The identity function is not continuous so $ f $ can't be that. For example if you take open set $ \{c\} $ in the codomain then its inverse should mean $ \{c\} $ is in the domain.

The function can be:

$ f(a) = c $

$ f(b) = a $ and by implication:

$ f(c) = b $

Then this is a bijection between the sets. Also it is continuous:

$ f^{-1}(c) = a $

$ f^{-1}(a) = b $ and by implication:

$ f^{-1}(b) = c $

We can say that the function induces a bijection:

$ \mathcal{T} \to \mathcal{U} $

$ \emptyset \to \emptyset $

$ \{a\} \to \{c\} $

$ \{a,b\} \to \{a,c\} $

$ X \to Y $

Therefore these two topological spaces are homeomorphic.
