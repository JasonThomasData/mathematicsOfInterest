### Galton board simulation

A Galton Board is a device which drops balls into a maze of dowels.

Each ball upon meeting a dowel has an approximately equal chance to go left or right.

By the time the balls settle at the bottom they form a binomial distribution.

The Central Limit Theorem says that if we had an infinite number of balls then the distribution would approach a perfect Normal Distribution.

```r
balls = 150
columns = 20
probability = 0.5

dist = rbinom(balls, columns, probability)
hist(dist)
```
