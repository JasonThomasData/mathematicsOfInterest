```octave
a_0 = 100000000;
b_0 = 1;

a_n = [a_0];
b_n = [b_0];

maxLoop = 20;

x = 1:maxLoop;

for i=1:maxLoop-1;
    n = length(a_n);
    a_np1 = sqrt(a_n(n) * b_n(n));
    b_np1 = (a_n(n) + b_n(n))/2;
    
    a_n = [a_n, a_np1];
    b_n = [b_n, b_np1];
end

plot(x, a_n);
hold on;
plot(x, b_n);
```
