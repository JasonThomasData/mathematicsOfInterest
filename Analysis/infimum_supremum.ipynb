### Infimum and supremum

These concepts will overlap with Topology.

Let $ A $ be a subset of the reals:

$ A \subset \mathbb{R} $ Note, can't be equal to $ \mathbb{R} $

$ x \in A $

$ A $ is bounded if:

$ M \in \mathbb{R} $

$ \forall x : M \ge x $

There might be a lot of $ M $ that exist but consider the smallest possible $ M $ that meets these criteria.

Then we can say that "the least possible $ M $ that bounds $ A $ is the supremum".

Let this be the notation:

$ m,M $ for any lower and upper bounds.

$ \sup $ or $ M'$  for the Supremum

$ \inf $ or $ m' $ for the Infimum.

#### Unbounded

$ A \subset \mathbb{R} $ is unbounded from above iff: $ \sup A = +\infty $ 

$ A \subset \mathbb{R} $ is unbounded from below iff: $ \inf A = -\infty $

#### $ \emptyset $ is unbounded

$ \sup \emptyset = -\infty $

$ \inf \emptyset = +\infty $

If it was the opposite with $ \sup \emptyset = +\infty $ then that would imply that something is in the set, but we know it's empty, so it can't be that way.

#### Infimum and supremum are unique

If they exist, then they are unique.

#### Infimum and supremum may or may not be in the set they bound

$ A = \{ x \in \mathbb{Q} : x < \sqrt{2} \} $

There is a proof that $ \sqrt{2} \not \in \mathbb{Q} $

- Any open set will not contain its $ \sup, \inf $

- Any closed set will not do so

- A hybrid, eg $ (0, 1] $ will contain its $ \sup $ but $ \inf $

#### Definition for $ \sup $

Let:

$ A \subset \mathbb{R}, A \ne \emptyset, M \in \mathbb{R} $, where $ M $ is an upper bound.

The Supremum is defined as:

$ M = \sup A $ __iff__:

$ \forall x \in A: x < M $ __AND__

$ \forall \epsilon > 0, \exists x' \in A, x' > M - \epsilon $

Where $ x' $ is the largest value of $ x $

#### Proof for the definition

$ (\Rightarrow) $

Suppose:

$ M = \sup A $

Then:

$ M $ is the smallest value that bounds $ A $ from above, and that implies $ \forall x \in A: x < M $

Also:

Let $ \epsilon $ be the positive limit approaching $ 0 $. Then $ x' > M - \epsilon $

$ (\Leftarrow) $

Suppose:

$ \forall x \in A: x < M $ __AND__

$ \forall \epsilon > 0, \exists x' \in A, x' > M - \epsilon $

Then:

If we consider the first supposition by itself, then $ M $ is an upper bound.

If we also consider the second supposition then take a particular $ \epsilon $ to be the positive limit approaching $ 0 $, then if some $ x' > M - \epsilon $ implies $ M $ must be the smallest possible value of any upper bound, and this is otherwise known as the $ M = \sup A $

#### $ \sup A $,  where $ A $ is monotonic and increasing, is the limit

Let:

$ A \subset \mathbb{R} $ be an increasing monotonic sequence.

Then if the supremum is not $ \infty $ then this means the set $ A $ is bounded. 

Actually, if the function is "not decreasing" then it's bounded above.

$ M = \sup x $

$ \{x_k \} < M $

So $ \forall x_k \le \sup x $

But: 

$\forall \epsilon > 0, \exists K(\epsilon), \forall k > K(\epsilon): x_k > M - \epsilon $

This is true:

$ x_k > M - \epsilon $

But so is this:

$ M + \epsilon > x_k $

And we can express those last two facts as a single statement:

$ | x_k - M | < \epsilon $

And then we put that back into the original statement:

$\forall \epsilon > 0, \exists K(\epsilon), \forall k > K(\epsilon): x_k > M - \epsilon $
