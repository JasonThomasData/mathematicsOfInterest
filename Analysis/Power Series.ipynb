### Power series

A power series is in the form:

$ \sum \limits_{n=0}^\infty c_n(x-a)^n $

$ = c_0 + c_1(x-a) + c_2(x-a)^2 + c_n(x-a)^3 $

$ a $ is the "centre of the power series"

The power series is a generalised geometric series:

$ a = 0 $

$ \sum \limits_{n=0}^\infty c_n(x-0)^n = \sum \limits_{n=0}^\infty c_n x^n$

Then a geometric series is the power series with its centre at 0.

You can think of a power series as a function of x of infinite order, and it can be a useful concept to derive the value of some well-known constants.

### Convergence of power series

There are 3 cases, and the only way I know how to do this is by using the ratio test.

#### case 1: The series converges if $ |x-a| < R $, diverges otherwise

In other words, if the absolute value of the distance $ x $ from the centre $ a $ is less than $ R $, then the series converges.

To find $ R $, evaluate $ \lim \limits_{n\to \infty} a_n $ via the ratio test.

Then remember that: $ |x-a| < R $, then the series converges.

$ -R < x-a < R $

$ -R+a < x < R+a $

This is the interval for which the series will converge.

#### case 2: $ x = a $, then the series converges trivially

This is the trivial case, and the series converges to $ 0 $ in this case.

eg: $ (x - a)^2 = 0^2 = 0 $ 

Think of this: 

$ R = 0 $

$ 0 < x-a < 0 $

Then the series converges _____

#### case 3: The series converges for all x

$ R = \infty $

$ -\infty < x-a < \infty $

### Finding new power series

#### By differentiating

Take the geometric series:

$ 1 + x + x^2 + x^3 + ... x^n $

This is of course the power series with a centre of $ a = 0 $.

Then the series converges:

$ | x | < 1 \implies \sum \limits_{n=0}^\infty x^n = \frac{1}{1-x} $

We can make a differential equation by differentiating both sides term by term.

$ \frac{d}{dx} ( 1 + x + x^2 + x^3 + ... x^n ) = \frac{d}{dx} \left( \frac{1}{1-x} \right) $

LHS = $ 0 + 1 + 2x + 3x^2 ... + nx^{n-1} $

LHS = $ \sum \limits_{n=0}^\infty (n+1)x^n $

RHS = $ \frac{d}{dx} \left( (1-x)^{-1} \right) $

RHS = $ -1 \cdot -1(x-1)^{-2} $

RHS = $ \frac{1}{1 - x^2} $

$ \therefore \sum \limits_{n=0}^\infty (n+1)x^n = \frac{1}{1 - x^2} $

### Evaluating functions of $ x $

#### Example 1

Let's find this series up until some degree, let's say 3:

$ \frac{\ln(1+2x)}{1+3x} $

If we remember that for an alternating geometric series:

$ |t| < 1 \implies \sum_{n=0}^\infty -t^n = \frac{1}{1+t} $

Then this is the same as:

$ \frac{1}{1+t} = 1 - t + t^2 - t^3 + ... $, while $ |t| < 1 $

Then we can integrate this to find:

$ \int \left( \frac{1}{1+t} \right) dt = \int \left(1 - t + t^2 - t^3 + ... \right) dt $

$ \ln(1+t) = t - \frac{t^2}{2} + \frac{t^3}{3} - ... $

But remember, the function in the example has the numerator: $ \ln(1+2x) $

Therefore, substitute in that value for $ x $

$ \ln(1+2x) = 2x - \frac{(2x)^2}{2} + \frac{(2x)^3}{3} - ... $

$ \ln(1+2x) = 2x - \frac{4x^2}{2} + \frac{8x^3}{3} - ... $

$ \ln(1+2x) = 2x - 2x^2 + \frac{8}{3}x^3 - ... $

Now we need to find:

$ \frac{1}{1+3x} $ 

And this already looks like an alternating geometric series:

$ \frac{1}{1+t} = 1 - t + t^2 - t^3 + ... $, while $ |t| < 1 $

But $ t = 3x $

$ \frac{1}{1+3x} = 1 - 3x + (3x)^2 - (3x)^3 + ... $

$ \frac{1}{1+3x} = 1 - 3x + 9x^2 - 27x^3 + ... $

The original problem was:

$ \frac{\ln(1+2x)}{1+3x} = \ln(1+2x)\frac{1}{1+3x}$

$ \frac{\ln(1+2x)}{1+3x} = (2x - 2x^2 + \frac{8}{3}x^3 - ...) (1 - 3x + 9x^2 - 27x^3 + ...) $

Keeping in mind that we only need to find the series up until the third degree, then we apply the distributive law, but ignore any result of $ x^n $ where $ x > 3 $

$ (2x*1 + 2x*-3x + 2x*9x^2) + (-2x^2 * 1 - 2x^2 * -3x) + (\frac{8}{3}x^3 * 1) $

$ (2x -6x^2 + 18x^3) + (-2x^2 + 6x^3) + (\frac{8}{3}x^3) $

$ 2x + (-6x^2 -2x^2) + (18x^3 + 6x^3 + \frac{8}{3}x^3) $

$ 2x - 8x^2 + \frac{80}{3}x^3 $

#### Example 2

Let's find the power series up to the 3rd degree.

$ \frac{1-4x}{1+2x} $

$ (1-4x)\frac{1}{1+2x} $

The right part above looks like the alternating geometric series:

$ \frac{1}{1-t} = t^0 + t^1 + t^2 + ... + t^n $

Except that $ t = -2x $

$ \frac{1}{1-(-2x)} = 1 -2x + 4x^2 -8x^3 + ... $

The part $ (1-4x) $ already is a polynomial, to the first degree, so:

$ \sum \limits{n=0}^\infty \frac{1-4x}{1+2x} = (1-4x)(1 -2x + 4x^2 -8x^3 + ...) $

$ (1*1 + 1*-2x + 1*4x^2 + 1*-8x^3) + (-4x*1 + -4x*-2x + -4x* 4x^2) $

$ (1 -2x + 4x^2 - 8x^3) + (-4x + 8x^2 -16x^3) $

$ (1 -6x + 12x^2 - 24x^3) $
