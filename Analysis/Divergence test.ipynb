### Divergence test

This is the first test to apply when you wish to determine if a series diverges.

Read:

$ \to $: tends to

$ \implies $: as implies

$ S = \sum \limits_{n=1}^\infty \left( a_n \right) $

The test:

$ \left\{ \lim \limits_{n \to \infty} a_n \ne 0 \right\} \implies S $ diverges

$ \left\{ \lim \limits_{n \to \infty} a_n = 0 \right\} \implies S $ test has no conclusion

The second case is inconclusive because the test is a contrapositive of this logic:

$ S_N = a_1 + a_2 + ... + a_{N-1} + a_N $

$ S_{N-1} =  a_1 + a_2 + ... + a_{N-1} $

$ S_N - S_{N-1} = a_N $

Both partial sums tend to the infinite series:

$ \lim \limits_{N \to \infty} S_N = S $

$ \lim \limits_{N \to \infty} S_{N-1} = S $

But also:

$ a_N = (S_N - S_{N-1}) \to (S - S) = 0 $

But that only makes sense if $ a_N $ converges, (because $ \infty $ minus anything is $ \infty $ but $ \infty - \infty $ is undefined.

$ \therefore $ this implication is important:

$ P \implies Q $:

$ S $ converges $ \implies a_N = 0 $

Then the contrapositive is:

$ \neg Q \implies \neg P $:

$ a_N \not \to 0 \implies S $ diverges.

But of course just $ Q $ wouldn't necessarily mean anything, so you need a seperate test for convergence.

#### Example: diverges

$ S = \sum \limits_{n=1}^\infty \left( \frac{n^2}{5n^2+4} \right) $

$ a_n = \frac{n^2}{5n^2+4} $

$ a_n = \frac{n^2}{n^2(5+4/n^2)} $

$ a_n = \frac{1}{5+4/n^2} $

$ \lim\limits_{n \to \infty} a_n = \frac{1}{5 + 0} $

Since $ a_n \not \to 0 $, then the series diverges.

#### Example: diverges (oscillates)

$ \sum \limits_{n=1}^\infty (-1)^n $

This diverges since:

$ a_1 + a_2 + a_3 + ... + a_n $

$ -1 + 1 -1 + ... + a_n $

This is a telescoping sum, if you remove the middle parts.

$ \sum \limits_{n=1}^\infty (-1)^n = (-1)^1 + (-1)^n $

This oscillates between -2 and 0, so therefore it will never converge.
