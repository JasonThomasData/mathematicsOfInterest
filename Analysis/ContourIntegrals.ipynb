### Example of a line segment

$ \gamma(t) $ is a line segment ending at $ 1 + i $

To write this as a function, we can start at $ z_1 $ and end at $ z_2 $:

$ (1 - t)z_1 + z_2t = 1 + i $

Where $ t = 0 $, the line is at the value $ z_1 $.

Where $ t = 1 $, the line is at the value $ z_2 $.

The integral is this:

$ \int_\gamma ze^{z^2} dz $

First find $ \frac{d}{dz} ze^{z^2} $:

$ \frac{d}{dz} (1+i)t e^{t(1+i)^2} $

$ = (1 + i)(1 + i)^2 e^{t(1+i)^2} $

$ = (4i -2)e^{t(1+i)^2} $

Then in form:

$ \int_a^b f(\gamma(t))\gamma'(t) dt $

$ \int_0^{1} (1+i)t e^{t(1+i)^2} \cdot (4i -2)e^{t(1+i)^2} dt $

$ \int_0^{1} (1+i)t (4i -2) e^{t(1+i)^2} dt $

$ \int_0^{1} (4it -2t -4t -2it) e^{t(1+i)^2} dt $

$ \int_0^{1} (2it -6t) e^{t(1+i)^2} dt $
