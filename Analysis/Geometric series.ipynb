### Geometric series

These are series where the difference between terms is constant.

EG:

$ a_n = \frac{1}{n^2} $

$ \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + ... + \frac{1}{n^2} $

Each term is half of the previous one, so this is a constant change, and therefore it's geometric.

#### Finding the limit of $r^n$ 

$ a_n = r^n $

$S_n =  \sum\limits_{n=0}^\infty r^n $

TLDR: if r < 1 then $ S_n $ tends to $\frac{1}{1-r} $. Otherwise $ S_n $ diverges.

Below, $ N - 1 $ because we start at 0, but need to still have $ N $ terms

$S_N =  \sum\limits_{n=0}^{N-1} r^{N-1} = r^0 + r^1 + r^2 + ... r^{N-1} $

$r S_N = r^1 + r^2 + r^3 + ... r^{N} $

$ r S_N - S_N = (r^0 + r^1 + r^2 + ... r^{N-1}) - r^1 + r^2 + r^3 + ... r^{N} $

$ r S_N - S_N = r^N - r^0 $

$  S_N(r - 1) = r^N - r^0 $

$  S_N(r - 1) = r^N - 1 $

$  S_N = \frac{r^N - 1}{ (r - 1) }, r \ne 1$

$  S_N = \frac{-1( -r^N + 1)}{ -1(-r + 1) }, r \ne 1$

$  S_N = \frac{-r^N + 1}{ -r + 1 }, r \ne 1$

#### A geometric series in disguise

$ S = \frac{3}{5} + \frac{3}{5^2} + \frac{3}{5^3} + ... $

This would need to have a 1 at the front for the former method to apply

$ S = \frac{3}{5} \left(1 + \frac{1}{5} + \frac{1}{5^2} + \frac{1}{5^3} + ... \right)$

$ \frac{3}{5} \sum \limits_{n=0}^\infty \left(\frac{1}{5}\right)^n $

$ r = \frac{1}{5} $

Since $ |r| < 1 $, then:

$ \sum \limits_{n=0}^\infty \left(\frac{1}{5} \right)^n = \frac{1}{1 - r} = \frac{1}{1 - 1/5} = \frac{1}{4/5} = \frac{5}{4} $

$ \frac{3}{5}\sum \limits_{n=0}^\infty \left(\frac{1}{5} \right)^n = \frac{3}{5} * \frac{5}{4} = \frac{3}{4}$

That is therefore a finite value, and it converges
