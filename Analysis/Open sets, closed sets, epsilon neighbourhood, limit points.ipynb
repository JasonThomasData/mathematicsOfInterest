### Epsilon neighbourhood in a metric space

When thinking about a $(\mathbb{R}^n, d) $ metric space then a good way to interpret the $ \epsilon $ that we would use for the real line is to imagine $ \epsilon $ is the radius of a ball

Given:

$ \bar x $ is the limit

$ (\mathbb{R}^n, d) $ is a metric space

Then

$ O_\epsilon(\bar x) = \{x \in \mathbb{R}^n, d(x,\bar x) < \epsilon \} $

### Closed sets

Should be __iff__

![closed sets definition](Screenshot_2023-07-30_22-54-01.png)

A set is closed if it contains all of its limit points.

#### Subsequences converge inside closed intervals

In a closed interval, a sequence is bounded above and below by points $ a, b $, which are constant sequences. If a sequence has only one number then it converges to itself. Since the sequence inside the closed interval is within two convergent sequences, then by the sandwich theorem, the sequence inside the bounded interval converges.

#### Borders and closed sets

Given a closed set $ S $, which is a set that contains its boundaries:

Then imagine there is a point $ \bar x $ on the border of $ S $. This is also a limit point.

Draw a ball around this point, where:

$ O(\bar x, \epsilon) \cap S \ne \emptyset $

Then since the set has an overlap with S, then $ \epsilon > 0 $.

Then we can use this fact to choose some $ K $ such that $ k \le K $, all points of $ x_k \subset S $

This is not true for open sets, because an open set doesn't contain its limit points, so there is no way for the limit of $ \bar x $ to sit on the border of an open set.
