### Cauchy integral formula

Let there be a disk $ D = \{ \gamma = re^{it} : \gamma > 0, t \in (0, 2\pi) \} $

And this disk could also have an absolute value representation.

As long as $ z $ is in the disk, and $ f $ is an analytic function, then you can represent this as:

$ f(a) = \frac{1}{2\pi i} \int_\gamma \frac{f(z)}{(z-a)} dz $

And this means that the $ D $ is a contour and we can apply the familiar notion of contour integrals.

#### Finding derivatives of the CIF

Given:

$ f(a) = \frac{1}{2\pi i} \int_\gamma \frac{f(z)}{(z-a)} dz $

$ f'(a) = \frac{d}{da} \left( \frac{1}{2\pi i} \int_\gamma \frac{f(z)}{(z-a)} dz \right) $

$ f'(a) = \frac{1}{2\pi i} \int_\gamma \frac{d}{da} \left( \frac{f(z)}{(z-a)} \right) dz $

$ f'(a) = \frac{1}{2\pi i} \int_\gamma \frac{f(z)}{(z-a)^2} dz $

And then the pattern for taking derivatives is:

$ f''(a) = \frac{1}{2\pi i} \int_\gamma 2 \frac{f(z)}{(z-a)^3} dz $

$ f'''(a) = \frac{1}{2\pi i} \int_\gamma 3 \frac{f(z)}{(z-a)^4} dz $

#### Generalised Cauchy Integral Formula (using derivatives)

Then generally, we get:

$ f^{(n)}(a) = \frac{n!}{2\pi i} \int_\gamma \frac{f(z)}{(z-a)^{n+1}} dz $

Where $ f^{(n)} $ means $ f $ to the $ nth $ derivative.

_But we must always remember, we assume $ f(z) $ is analytic and that the point $ a $ is inside the contour_

#### Example with $ a $ as interior point of $ \gamma $

$ f(a) = \int_\gamma \frac{e^z}{z-2} dz $

And let:

$ \gamma = \{|z| = 3\} $, which is the circle at the origin of radius $3$.

Then we can identify $ f(z) = e^z, a = 2 $

Then we can observe that $ f(z) $ is analytic and $ a $ is in the interior of $ \gamma $. That means we can use Cauchy's Integral Formula.

$ f(a) = \frac{1}{2\pi i} \int_\gamma \frac{e^z}{z-2} dz $

But $ f(z) = e^z $

So $ f(a=2) = e^2 $

So
 
$ e^2 = \frac{1}{2\pi i} \int_\gamma \frac{e^z}{z-2} dz $

$ e^2 2 \pi i = \int_\gamma \frac{e^z}{z-2} dz $

#### Example with $ a $ not an interior point of $ \gamma $

$ f(a) = \int_\gamma \frac{e^z}{z-2} dz $

And let:

$ \gamma = \{|z| = 1\} $, which is the circle at the origin of radius $ 1 $.

Then we can identify $ f(z) = e^z, a = 2 $

Then we can observe that $ f(z) $ is analytic and $ a $ is not in the interior of $ \gamma $. 

But since the integrand is analytic, and since the contour is closed, then the integral must equal 0.

So $ f(a) = 0 $

##### Summary of above two examples

- If the formula has a closed contour and the point $ a $ is in the interior, and if $ f(z) $ in the numerator is analytic, then find the integral using the Cauchy Integral Formula.
- If the formula has a closed contour and the point $ a $ is outside, so long as the entire function is analytic, it has integral of $ 0 $.

#### Example with fraction decomposition

$ f(a) = \int_\gamma \frac{e^z \sin z}{z^2-9} dz $

$ \gamma = 4e^{it}, t \in [0, 2\pi] $

$ f(z) $ is analytic but the denominator needs to be expressed as single powers of $ z $

We can find partial fractions.

$ \frac{1}{z^2 - 9} = \frac{1}{(z+3)(z-3)} $ 

$ \frac{1}{z^2 - 9} = \frac{A}{(z+3)} + \frac{B}{(z-3)} $ 

$ 1 = A(z-3) + B(z+3) $ 

The as usual, you can set z so that one of A or B disappears.

$ z = 3 $

$ 1 = B6 $

$ B = \frac{1}{6} $

$ z = -3 $

$ 1 = -A6 $

$ A = -\frac{1}{6} $

$ \frac{1}{z^2 - 9} = \frac{1}{6} \left[\frac{1}{(z-3)} - \frac{1}{(z+3)} \right] $ 

So now you substitute the analytic function back in and you get:

$ \frac{e^z \sin z}{z^2 - 9} = \frac{1}{6} \left[\frac{e^z \sin z}{(z-3)} - \frac{e^z \sin z}{(z+3)} \right] $ 

$ \frac{1}{6} \left[ \int_{\gamma} \frac{e^z \sin z}{z-3} dz - \int_{\gamma} \frac{e^z \sin z}{z+3} dz \right] $

Now we can apply the CIF, and remember the sign of $ a $ in the denominator.

$ \frac{1}{6}[f(3) - f(-3)] = \frac{1}{6} \left[\frac{1}{2 \pi i} \int_{\gamma} \frac{e^z \sin z}{z-3} dz - \frac{1}{2 \pi i} \int_{\gamma} \frac{e^z \sin z}{z+3} dz \right] $

$ \frac{1}{6}[f(3) - f(-3)] = \frac{1}{6} \frac{1}{2 \pi i} \left[\int_{\gamma} \frac{e^z \sin z}{z-3} dz - \int_{\gamma} \frac{e^z \sin z}{z+3} dz \right] $

$ \frac{1}{6}[e^{3}\sin 3 - e^{-3}\sin -3] = \frac{1}{6} \frac{1}{2 \pi i} \left[\int_{\gamma} \frac{e^z \sin z}{z-3} dz - \int_{\gamma} \frac{e^z \sin z}{z+3} dz \right] $

$ \frac{\pi i}{3}[e^{3}\sin 3 - e^{-3}\sin -3] = \frac{1}{6} \left[\int_{\gamma} \frac{e^z \sin z}{z-3} dz - \int_{\gamma} \frac{e^z \sin z}{z+3} dz \right] $

The minus in the $ \sin $ and the minus in front of that term cancel:

$ \frac{\pi i}{3}[e^{3}\sin 3 + e^{-3}\sin 3] $

$ \frac{\pi i}{3} (\sin 3) [e^{3} + e^{-3}] $

$ \frac{2\pi i}{3} (\sin 3) [\frac{e^{3} + e^{-3}}{2}] $

$ \frac{2 \pi i}{3} \sin 3 \cosh 3 $
