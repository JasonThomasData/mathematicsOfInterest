### Limit of a function

The limit is:

$ b = \lim_\limits{x \to a} \left( f(x) \right) $

If a limit exists then:

$ | x_k - a | < \delta \to | y_k - b | < \epsilon $

Then given some $\epsilon$ we need to find $ \delta $

#### Basic example without bounding

$ f(x_k) = e^{1/x_k} $

Here's the graph:

PIC

Then we can guess that this goes to a limit of $ 0 $

$ |y_k - b | < \epsilon $

$ |y_k| < \epsilon $

Then that's just the function itself, so:

$ e^{1/x_k} < \epsilon $

To find $ \delta $:

$ \ln e^{1/x_k} < \ln \epsilon $

$ \frac{1}{x_k} < \ln \epsilon $

$ \frac{1}{|x_k|} > - \ln \epsilon $

$ \frac{1}{-\ln \epsilon} > |x_k|$

$ |x_k| < \frac{-1}{\ln \epsilon} $

Now we have a function of $ x $ on the LHS and $ \epsilon $ on the right hand side, so we can set $ \delta $ to be this LHS value.

Then this is equivalent to saying:

$ |x_k| < \delta = \frac{-1}{\ln \epsilon} $

#### Example with bounding

$ f(x) = \lim_\limits{x \to 0} \frac{x^2}{x-1} \cdot \cos \frac{1}{x} = 0 $ 

Then to prove the above, we can use the absolute value theorem:

If this converges to $ 0 $:

$ \left| \frac{x^2}{x-1} \cdot \cos \frac{1}{x} \right| $

Then so does the function given originally.

We can bound $ \left| \cos \frac{1}{x} \right| $ from above because $ | \cos \theta |$ oscillates between $ [0,1] $

So now we consider a reduced supposed limit:

$ \left| \frac{x^2}{x-1} \right| = \frac{x^2}{|x-1|} = 0 $

Then it would be good to bound the denominator.

We can do it with the reverse triangle inequality:

$ |a - b| \ge \large\left|\small |a| - |b| \large \right| \small $

$ |x - 1| \ge \large\left|\small |x| - 1 \large \right| \small $

$ |x - 1| \ge 1 - |x| $

If $ |x| < \frac{1}{2} $ then $ 1 - |x| \ge \frac{1}{2} $

$ |x - 1| \ge \frac{1}{2} $

Then note that a larger denominator makes a number smaller.

So then this is with the denominator bounded:

$ \frac{x^2}{|x-1|} \le \frac{x^2}{1/2} $

$ \frac{x^2}{|x-1|} \le 2x^2 $

So now we have bounded the function above, so let's now reintroduce $ \epsilon $

$ 2x^2 < \epsilon $

$ x^2 < \frac{\epsilon}{2} $

$ |x| < \sqrt{\frac{\epsilon}{2}} $

But if we reintroduce the supposed limit:

$ | x - 0 | < \sqrt{\frac{\epsilon}{2}} $

Then this looks very much like:

$ | x - 0 | < \delta $

So this is a sufficient value for $ \delta $
