### Bivariate random variables

#### Discrete, joint pmf

The r.v. $ (X, Y) $ is discrete if $ X $ and $ Y $ are discrete.

The pmf of a discrete bivariate r.v. is $ p(x, y) = P(X = x, Y = y) $. Likewise you can compute this as a subset.

This must be true for $ (x, y) $:

$ \sum_x \sum_y p(x, y) = 1 $

#### Discrete, joint cdf

In this notes read $ \infty $ as $ \lim_\limits{x\to \infty} $

i)

If either $ x, y $ are impossible then $ F(x, y) = 0 $:

$ F(-\infty, y) = 0 $

$ F(x, -\infty) = 0 $

$ F(-\infty, -\infty) = 0 $

ii)

$ 0 \le F(x, y) \le 1 $ 

iii)

$ F(\infty, \infty) = 1 $ 

iv)

Let: $ x_1, x_2 $ be limits of $ X $ and $ y_1, y_2 $ be limits of $ Y $

Then if you want to know the probability of a bivariate r.v., example weight height in a range. Then what's the probability of the joint event in terms of weight and height

$ P(x_1 \le X \le x_2, y_1 \le Y \le y_2) = F(x_1, x_2) + F(x_2, x_2) - F(x_1, y_2) - F(x_2, y_1) $

#### Marginal CDFs (discrete)

$ F_x(x) = P(X \le x) = F(x, \infty) $, for all $ y $

$ F_y(y) = P(Y \le y) = F(\infty, y) $, for all $ x $

But if $ X, Y $ takes a finite number of possible values then:

$ F_x(x) = P(X \le x) = F(x, \max(y)) $, for all $ y $

$ F_y(y) = P(Y \le y) = F(\max(x), y) $, for all $ x $

#### Marginal PMFs (discrete)

$ P(X = x) = p_x(x) = \sum_y p(x, y) $ (sum over all y's) 

$ P(Y = y) = p_y(y) = \sum_x p(x, y) $ (sum over all x's)

#### Example finding joint discrete cdf (from a pmf), and then finding marginal cdfs

$ p(x, y) = \left(\frac{2}{3}\right)^{x+y} \left(\frac{1}{3}\right)^{2-x-y} , x \in \{0, 1\}, y \in \{0, 1\} $

$ p(0, 0) = \left(\frac{2}{3}\right)^{0+0} \left(\frac{1}{3}\right)^{2-0-0} = \frac{1}{9} $

$ p(1, 0) = \left(\frac{2}{3}\right)^{1+0} \left(\frac{1}{3}\right)^{2-1-0} = \frac{2}{9} $

$ p(0, 1) = \left(\frac{2}{3}\right)^{0+1} \left(\frac{1}{3}\right)^{2-0-1} = \frac{2}{9} $

$ p(1, 1) = \left(\frac{2}{3}\right)^{1+1} \left(\frac{1}{3}\right)^{2-1-1} = \frac{4}{9} $

They all sum up to $ 1 $, so this is a valid pmf

Now we find the marginal cdfs, remember $ F $ is cumulative and adds up to $ 1 $:

$ F_x(0) = F(0, \infty) = F(0, 1) = p(0, 0) + p(0, 1) = \frac{1}{3} $

$ F_x(1) = F(1, \infty) = F(1, 1) = p(0, 0) + p(0, 1) + p(1, 0) + p(1, 1) = 1 $

$ F_y(0) = F(\infty, 0) = F(1, 0) = p(0, 0) + p(1, 0) = \frac{1}{3} $

$ F_y(1) = F(\infty, 1) = F(1, 1) = p(0, 0) + p(0, 1) + p(1, 0) + p(1, 1) = 1 $

Above, $ \infty \to 1 $ because $ \max{x} = 1, \max{y} = 1 $

### Mutual exclusive

If a pdf or pmf of a bivariate r.v. can be written as a product of functions (in this case they're marginal probabilities):

$ p(x, y) = p_x(x) p_y(y) $

$ f(x, y) = f_x(x) f_y(y) $

Then $ X, Y $ are mutually exclusive. This extends to have more r.v.

What you have to show is that the joint pmf/pdf is a product of the marginals.



#### Continuous, joint probability (pdf)

Assume we already have a jount pdf $ f(x, y) $

To obtain the joint probability, we do:

$ P(x_1 \le X \le x_2, y_1 \le Y \le y_2) = \int_{y_1}^{y_2} \int_{x_1}^{x_2} f(x, y) \space dx \space dy $

$ f(x, y) \le 0 $ for all pairs

$ \int_{y_1}^{y_2} \int_{x_1}^{x_2} f(x, y) \space dx \space dy = 1 $

#### Continuous, joint cdf

To obtain the joint cdf, we can use a dummy variable $ t_1, t_2 $:

$ F(x, y) = \int_{-\infty}^{x} \int_{-\infty}^{y} f(t_1, t_2) \space dt_1 \space dt_2 $

#### Continuous, marginal cdf

Instead of summing as with discrete r.v. we must integrate.

$ f_x(x) = \int_{-\infty}^\infty f(x, y) \space dy $

$ f_y(y) = \int_{-\infty}^\infty f(x, y) \space dx $

### Conditional probability

If two r.v.s are not independent then we can express the joint pmf/pdf as a condition probabilit/density

Recall the definition for conditional probability:

$P(A|B) = \frac{(A \cup B)}{P(B)} $

$ (A \cap B) = P(A|B)P(B) = P(B|A)P(A) $

Let:

$ A = \{ X = x \} $

$ B = \{ Y = y \} $

Then:

$ p(x|y) = \frac{p(x, y)}{p(y)} $, where $ p(y) \ne 0 $, for a discrete r.v.

$ f(x|y) = \frac{f(x, y)}{f(y)} $, where $ f(y) \ne 0 $, for a continuous r.v.

### Continuous r.v. expected value

Let:

$ g(x, y) $ be any function

$ f(x, y) $ is the pdf (given)

$ \mu = E(g(x, y)) = \int_{-\infty}^\infty \int_{-\infty}^\infty g(x, y) f(x, y) dx \space dy $

### Continuous? r.v. covariance

Not sure if this relates to discrete r.v.s, but definitely is for continuous.

This is a particular case of the expected value, above, where we want to find the covariance

Let:

$ g(x, y) = (x - E(X)(y - E(Y) $

$ f(x, y) $ is the pdf (given)

$ cov(x, y) = \int_{-\infty}^\infty \int_{-\infty}^\infty (x - E(X)(y - E(Y ) f(x, y) dx \space dy $

And then there is another way:

$ cov(X, Y) = E(XY) - E(X)E(Y) $

_If $ X, Y $ are independent r.v.s, then $ cov(x, y) = 0 $. The inverse is generally not true._

### Continuous? r.v. correlation

$ \rho = \frac{cov(x, y)}{\sigma_x \sigma_y} $

$ \rho = 0 $ implies no correlation.

Lack of correlation does not imply independence.

### Example - bivariate expected value, covariance, correlation

$ f(x, y) = \left\{ \begin{matrix} \frac{(x+y)^2}{40} & , &  -1 < x < 1, -3 < y < 3 \\ 0 & , & \space \text{otherwise} \end{matrix} \right\} $

$ m_{11} $ is the moment where: $ \mu = E(XY) $

$ m_{10} $ is the moment where: $ \mu_X = E(X) $

$ m_{01} $ is the moment where: $ \mu_Y = E(Y) $

##### What is $ m_{11} $ ?

$ E(XY) = \int_{-3}^3 \int_{-1}^1 xy f(x, y) \space dx \space dy $

$ E(XY) = \int_{-3}^3 \int_{-1}^1 xy \frac{(x+y)^2}{40} \space dx \space dy $

##### What is $ m_{20} $ ?

$ E(X^2Y^0) = \int_{-3}^3 \int_{-1}^1 x^2 y^0 f(x, y)\space dx \space dy $

##### What is $ m_{02} $ ?

$ E(X^0Y^2) = \int_{-3}^3 \int_{-1}^1 x^0 y^2 f(x, y)\space dx \space dy $

##### What is the variance of $ X $?

$ \sigma_x^2 = E(X^2) - E(X)^2 $

$ \sigma_x^2 = E(m_{20}) - E(m_{10})^2 $

$ m_{10} = E(X^1 Y^0) = \int_{-3}^3 \int{-1}^1 x^1 y^0 \space dx \space dy $

The variance of $ Y $ is the same as above, but you switch the position of the indices.

##### What is the correlation coefficient?

This is given in the notes above:

$ cov(X, Y) = E(XY) - E(X)E(Y) $

Then:

$ cov(X, Y) = E(X^1 Y^1) - E(X^1 Y^0)E(X^0 Y^1) $

$ cov(X, Y) = E(m_{11}) - E(m_{10})E(m_{01}) $

Then do:

$ \rho = \frac{cov(X, V)}{\sigma_x \sigma_y} $

$ \rho = \frac{E(m_{11}) - E(m_{10})E(m_{01})}{ \sqrt{E(m_{20}) - E(m_{10})^2} \sqrt{E(m_{02}) - E(m_{01})^2}} $

### Linear combinations of expected value and variance

If you have r.v.s $ X_1, X_2, ... X_n $

$ \sigma_X^2 = a_1^2 \sigma_{X_1}^2 + a_2^2 \sigma_{X_2}^2 + ... + a_n^2\sigma_{X_n}^2 $

$ E(X) = a_1 E(X_1) + a_2 E(X_2) + ... + a_n E(X_n) $

### Conditional expectation

The conditional expectation of $ Y $ given some particular value of $ X = x $ is $E(Y|x)$

If we haven't specified a value of $ X $ then we write it as $ E(Y|X) $

Generally:

$ E(Y) = E(E(Y|X))$

And this result means that when you find the conditional probability that $ X $ disappears, either by addition or integration. Therefore the second expected value is just for $ Y $.

#### Discrete r.v.

$ E(Y|x) = \sum_y y p(y|x) $

#### Continuous r.v.

$ E(Y|x) = \int y f(y|x) dy $

### Conditional variance

The conditional variance of $ Y $ given some particular value of $ X = x $ is $ Var(Y|x)$

If $ X $ is unknown then $ E(Y|X) $

Recall that $ Var(X) = E(X^2) - E(X)^2 $

#### Discrete r.v.

$ Var(Y|x) = \sum_y y^2 p(y|x) - (\sum_y y p(y|x))^2 $

#### Continuous r.v.

$ Var(Y|x) = \int_y y^2 f(y|x)\space dy - (\int_y y f(y|x) dy)^2 $

#### Useful for advanced work

$ Var(Y) = E(Var(Y|X)) + Var(E(Y|X)) $

### Example - finding conditional cdf, conditional expectation, conditional variance, starting from a joint pdf

_There is a similar example that focuses on order statistics in unit 3 lecture iic_

Given some joint cdf (continuous in this example):

$ f(x, y) $

This joint pdf might have ranges. Those will need to propagate through all steps here. I think the range for each applies to every step and should be calculated at every step.

First find the marginal densities: $ f_x(x), f_y(y) $

At this point, you can determine if $ X, Y $ are independent:

$ f(x, y) = f(x)f(y) $ will imply they are independent.

Then use those to find conditional probabability:

$ f_y(y|x) = \frac{f(x, y)}{f(x)} $, and same for $ f_x $

Now you can use that to find the conditional expectation: 

$ E(Y|x) = \int y f(y|x) dy $

And you can use that to find the conditional variance

$ Var(Y|x) = \int_y y^2 f(y|x)\space dy - (\int_y y f(y|x) dy)^2 $

Finally, we can find the expected value of just $ Y $:

$ Var(Y) = E(E(Y|X)) $

And again, don't forget to substitute in the range of $ x, y $ for every step. An example is given at the lecture week3 ii b.
