### Uniform distribution

#### PMF Discrete (find probability)

$ p(x_i) = \frac{1}{n} $

OR

$ p(x_i) = 0 $

#### PDF Continuous (find probability)

$ f(x) = \frac{1}{b - a} $ if $ a < x < b $

otherwise

$ f(x) = 0 $ 

#### Expected (mean) value (discrete)

$ E(X) = \frac{n - 1}{2} $

This is because of Gauss' formula for addition:

$ \sum_\limits{i=1}^n = \frac{n(n-1)}{2} $

But since $ p(x_i) = \frac{1}{n} $

Then the probability of all $ x_i \in X $ is:

$ \frac{1}{n}\sum_\limits{i=1}^n = \frac{1}{n}\frac{n(n-1)}{2} $

#### Expected (mean) value (continuous)

$ \frac{a + b}{2} $

#### Cumulative distribution function (continuous)

If $ x < a $ then $ F(x) = 0 $

If $ a < x < b $ then:

$ F(x) = \int_a^x f(t) dt = \int_a^x \frac{1}{b - a} dt = \frac{x - a}{b - a} $

If $ b < x $ then $ F(x) = 1 $
