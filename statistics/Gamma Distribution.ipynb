### Gamma Distribution

This is a versatile distribution.

$ X \sim \Gamma(\alpha, \beta) $

Where:

$ \alpha $ is the shape parameter.

$ \beta $ is the scale parameter.

#### Expected value

$ E(X) = \alpha \beta $

#### Variance

$ Var(X) = \alpha \beta^2 $

#### Exponential distribution

When $ \alpha = 1 $ then the $ \Gamma $ distribution becomes the exponential distribution. The exponential distribution in this case takes $ \frac{1}{\beta} $ as a parameter.

#### Recursive relationship

I have not written down the $ \Gamma $ function itself.

The lecturer ensures this is a useful fact.

$ \Gamma(\alpha) = (\alpha -1)\Gamma(\alpha - 1) $

$ \Gamma(\alpha) = (\alpha -1)(\alpha -2)\Gamma(\alpha - 2) $

$ \Gamma(\alpha) = (\alpha -1)(\alpha -2)(\alpha -3)(\alpha -3) $ etc...

This stops at $ \Gamma(1) $ and we know that the pdf $ f(x) = 0 $, when $ x \le 0 $

Then the result can be summarised as:

$ \Gamma(\alpha) = (\alpha - 1)! $

Therefore:

$ \Gamma(1) = (1-1)! = 0! = 1 $

$ \Gamma(2) = (2-1)! = 1! = 1 $

$ \Gamma(3) = (3-1)! = 2! = 2 $

$ \Gamma(4) = (4-1)! = 3! = 6 $ etc

#### Parameter of 1/2

A result of the Gaussian Integral is that:

$ \Gamma \left(\frac{1}{2}\right) = \sqrt{\pi} $
