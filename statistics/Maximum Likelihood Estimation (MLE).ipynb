### Maximum Likelihood Estimation

This is useful when the form of the density (and also mass?) function is known but the parameter for the distribution is not known.
    
Steps:
- Identify pmf/pdf for random sample
- Find likelihood function $ L(\theta; X) $
- Find log-likelihood function $ l(\theta; X) $ (usually easier to differentiate)
- Differentiate:

$ \frac{dl(\theta; X_1, X_2, X_3, ... X_n )}{d\theta}|_{\theta = \hat \theta} = 0 $

- Solve that to find the MLE
- Find the second order derivative to ensure that it is the maximum

#### Example of Poisson Distribution

$ p(X, \theta) = \frac{\theta^x e^{-\theta}}{x!} , x = \{0,1,2,3,4\} $

$ L(\theta; X_1, X_2, ..., X_n) = \large \prod_{i=1}^n \frac{\theta^x e^{-\theta}}{x!} $

$ L(\theta; X_1, X_2, ..., X_n) = \large \frac{\prod_{i=1}^n \theta^x \prod_{i=1}^n e^{-\theta}}{ \prod_{i=1}^n x!} $

$ L(\theta; X_1, X_2, ..., X_n) = \large \frac{\theta^{(\sum_{i=1}^n x)} e^{-\theta n}}{ \prod_{i=1}^n x!} $

$ l(\theta; X_1, X_2, ..., X_n) = \ln \left[ \large \frac{\theta^{(\sum_{i=1}^n x)} e^{-\theta n}}{ \prod_{i=1}^n x!} \right] $

$ l(\theta; X_1, X_2, ..., X_n) = \ln \left[ \large \frac{\theta^{(\sum_{i=1}^n x)} e^{-\theta n}}{ \prod_{i=1}^n x!} \right] $

$ l(\theta; X_1, X_2, ..., X_n) = \ln \left[ \large \theta^{(\sum_{i=1}^n x)} \right] + \ln \left[ e^{-\theta n} \right] - \ln \left[ \prod_{i=1}^n x! \right] $

$ l(\theta; X_1, X_2, ..., X_n) = (\ln \theta) (\sum_{i=1}^n x) -\theta n - \ln \left[ \prod_{i=1}^n x! \right] $

That's about as simple as it gets, let's differentiate:

$ \frac{dl}{d\theta} = \frac{(\sum_{i=1}^n x)}{\theta} - n$

$ \frac{d^2l}{d\theta^2} = -(\sum_{i=1}^n x)\theta^{-2} $

To find the MLE $ \hat \theta $, concentrate on the first derivative, and evaluate at $ \theta = \hat \theta $ where the expression equals $ 0 $.

$ \frac{dl}{d\theta} = (\sum_{i=1}^n x)\frac{1}{\hat \theta} - n = 0$

$ \frac{dl}{d\theta} = (\sum_{i=1}^n x)\frac{1}{\hat \theta} = n $

$ \frac{dl}{d\theta} = (\sum_{i=1}^n x)\frac{1}{n} = \hat \theta $

Then this, above, is the mean or expected value for the distribution.

Let's also prove that this is the maximum, also by evaluating that the second derivative at $ \theta = \hat \theta $, is less than $ 0 $

$ \frac{d^2l}{d\theta^2} = -(\sum_{i=1}^n x) \hat \theta^{-2} < 0 $

It is certainly true that $ (\sum_{i=1}^n x) \ge 0 $, because the index of $ x_i $ are natural numbers.

Since the function uses real values, then $ \frac{1}{\hat \theta^2} \ge 0 $. The second derivative is a negative sign with two positive numbers, so the expression is $ < 0 $
