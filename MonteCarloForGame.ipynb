```python
# Now let's play a game

import random

cards = locationCards + locationCards + scenicCards + boonCards + npcEncounterCards + enemyEncounterCards
random.shuffle(cards)

npcGoodwill = {
    "Submarine": 0,
    "Supply boat": 0,
    "Platform workers": 0,
    "Rescue divers": 0,
    "Trapped survivors" : 0,
    "Total": 0
}

playerState = {
    "Energy": {"max": 7, "current": 3},
    "Hull integrity": {"max": 7, "current": 3},
    "Submerged": True
}

faceUpDiscardPile = {
    "Enemy encounter": [],
    "Location": [],
    "Npc encounter": []
}

def resolveAction(outcomes, playerState, action, goodwill, goodwillRatio):
    if action not in outcomes.keys():
        raise Exception("{} not in keys for outcomes: {}".format(action, outcomes.keys()))
    if goodwill > 0 and goodwillRatio is None:
        raise Exception("goodwillRatio not defined")

    if goodwill > 0:
        playerState["Energy"]["current"] += outcomes[action]["Energy"] + round(goodwill*goodwillRatio["Energy"])
        playerState["Hull integrity"]["current"] += outcomes[action]["Hull integrity"] + round(goodwill*goodwillRatio["Hull integrity"])
    else:
        playerState["Energy"]["current"] += outcomes[action]["Energy"]
        playerState["Hull integrity"]["current"] += outcomes[action]["Hull integrity"]
        
def drawCard(cards):
    card = cards.pop(0)

    cardType = card["Type"]
    print(cardType)

    if cardType == "Location":
        print("Surface:", card["Surface"]["Name"])
        print("Submerged:",card["Submerged"]["Name"])
        print("Choose from Scavenge, Repair, Thrust, or Rest")
    elif cardType == "Npc encounter":
        print(card["Name"])
        print("Level:", card["Level"])
        print("Choose from Give help, Request help, Ignore")
    elif cardType == "Boon":
        print(card["Name"])
        print("Receive reward")
    elif cardType == "Enemy encounter":
        print(card["Name"])
        print("Level:", card["Level"])
        print("Choose from Ram, Flee")
    
    return card

def goToLevel(level):
    if level == "Submerged":
        playerState["Submerged"] = True
    else:
        playerState["Submerged"] = False
        
def takeAction(card, action, goodwillRatio = None):
    if action is None:
        return

    if playerState["Submerged"]:
        level = "Submerged"
    else:
        level = "Surface"

    goodwill = 0

    if card["Type"] == "Npc encounter":
        goodwill = npcGoodwill[card["Name"]]
        if action == "Give help":
            npcGoodwill[card["Name"]] += card["Outcomes"]["Goodwill"]["Positive"]
            npcGoodwill["Total"] += card["Outcomes"]["Goodwill"]["Positive"]
        elif action == "Request help":
            npcGoodwill[card["Name"]] += card["Outcomes"]["Goodwill"]["Negative"]
            npcGoodwill["Total"] += card["Outcomes"]["Goodwill"]["Negative"]
        outcomes = card["Outcomes"]
    elif card["Type"] == "Location":
        outcomes = card[level]["Outcomes"]
    elif card["Type"] == "Scenic":
        return
    else:
        outcomes = card["Outcomes"]

    resolveAction(outcomes, playerState, action, goodwill, goodwillRatio)
    

def resolveBoon(card, energy, hull):
    playerState["Energy"]["current"] += energy
    playerState["Hull integrity"]["current"] += hull
    
def addCardToDiscardPile(card):
    if card["Type"] == "Location":
        if playerState["Submerged"]:
            name = card["Submerged"]["Name"]
        else:
            name = card["Surface"]["Name"]
    else:
        name = card["Name"]
        
    faceUpDiscardPile[card["Type"]].append(name)

def log(logList, cardName, action, playerState):
    logList.append({"cardName": cardName, "action": action, "playerState": str(playerState)})
    
def isPlayerDead(playerState):
    return (playerState["Energy"]["current"] <= 0 or playerState["Hull integrity"]["current"] <= 0)
```

```python
# Random player logic

def randomPlayerGame(cardsToDraw):

    cards = locationCards + locationCards + scenicCards + boonCards + npcEncounterCards + enemyEncounterCards
    random.shuffle(cards)

    npcGoodwill["Submarine"] = 0
    npcGoodwill["Supply boat"] = 0
    npcGoodwill["Platform workers"] = 0
    npcGoodwill["Rescue divers"] = 0
    npcGoodwill["Trapped survivors"] = 0
    npcGoodwill["Total"] = 0

    playerState["Energy"]["current"] = 3
    playerState["Hull integrity"]["current"] = 3
    playerState["Submerged"] = True

    faceUpDiscardPile["Enemy encounter"] = []
    faceUpDiscardPile["Location"] = []
    faceUpDiscardPile["Npc encounter"] = []
    
    logList = []
    scenicCount = 0
    thrustOn = False

    for i in range(cardsToDraw):
        if isPlayerDead(playerState):
            #print(logList)
            return len(logList), scenicCount

        card = cards.pop(0)

        #print(playerState)
        #print(card["Type"])

        if playerState["Submerged"]:
            level = "Submerged"
        else:
            level = "Surface"

        cardType = card["Type"]

        if cardType == "Scenic":   
            scenicCount += 1
        elif cardType == "Location":
            if random.randint(0,1) == 0:
                playerState["Submerged"] = not playerState["Submerged"]
            possibleActions = ["Scavenge", "Repair", "Thrust", None]
            randomAction = random.choice(possibleActions)
            takeAction(card, randomAction)
            log(logList, card[level]["Name"], randomAction, playerState)
            addCardToDiscardPile(card)
        elif cardType == "Enemy encounter" and (playerState["Submerged"] == card["Level"] or card["Level"] == "All"):
            if thrustOn:
                thrustOn = False
                log(logList, card, "Thrusted to bypass", playerState)
                continue
            else:
                possibleActions = ["Ram", "Flee"]
                randomAction = random.choice(possibleActions)
                if randomAction == "Ram":
                    addCardToDiscardPile(card)
                takeAction(card, randomAction)
                log(logList, card["Name"], randomAction, playerState)
        elif cardType == "Npc encounter" and (playerState["Submerged"] == card["Level"] or card["Level"] == "All") :
            name = card["Name"]
            goodwill = npcGoodwill[name]
            if goodwill >= 0:
                energyRatio = random.randint(0, goodwill)
                goodwillRatio = {"Energy": energyRatio, "Hull integrity": goodwill-energyRatio}
                possibleActions = ["Give help", "Request help", "Ignore"]
                randomAction = random.choice(possibleActions)
                if randomAction != "Ignore":
                    addCardToDiscardPile(card)
                    takeAction(card, randomAction, goodwillRatio)
                log(logList, card["Name"], randomAction, playerState)
            else:
                possibleActions = ["Give help", "Ignore"]
                randomAction = random.choice(possibleActions)
                if randomAction != "Ignore":
                    addCardToDiscardPile(card)
                    takeAction(card, randomAction)
                log(logList, card["Name"], randomAction, playerState)
        elif cardType == "Boon":
            if card["Name"] == "Robotic military drone":
                enemiesKilled = len(faceUpDiscardPile["Enemy encounter"])
                energyBoon = random.randint(0, enemiesKilled)
                hullBoon = enemiesKilled - energyBoon
                resolveBoon(playerState, energyBoon, hullBoon)
            elif card["Name"] == "Emergency rescue sub":
                if npcGoodwill["Total"] > 3:
                    if random.randint(0,1) == 0:
                        resolveBoon(playerState, 1, 0)
                    else: 
                        resolveBoon(playerState, 0, 1)
            elif card["Name"] == "Submarine seeks information":
                if len(faceUpDiscardPile["Locations"]) >= 4:
                    if random.randint(0,1) == 0:
                        resolveBoon(playerState, 1, 0)
                    else: 
                        resolveBoon(playerState, 0, 1)
            elif card["Name"] == "Ray rider offers a gift":
                rayRidersKilled = 0
                for killedName in faceUpDiscardPile["Enemy encounter"]:
                    if killedName == "Ray Riders":
                        rayRidersKilled += 1
                if rayRidersKilled < 2:
                    if random.randint(0,1) == 0:
                        resolveBoon(playerState, 1, 0)
                    else: 
                        resolveBoon(playerState, 0, 1) 

    return len(logList), scenicCount

actionCardTally = []
scenicCardTally = []
for j in range(100):
    actionCount, scenicCount = randomPlayerGame(60) 
    actionCardTally.append(actionCount)
    scenicCardTally.append(scenicCount)

print(sum(actionCardTally)/len(actionCardTally))
print(sum(scenicCardTally)/len(scenicCardTally))
```
