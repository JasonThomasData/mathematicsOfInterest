```python
from graphviz import Digraph
import math
```

```python
labels = ["a", "b", "c"]
transition_matrix = [
    [0.7, 0.2, 0.1],
    [0.5, 0.2, 0.3],
    [0, 1, 0]
]
```

```python
n_rows = len(transition_matrix)
for row in transition_matrix:
    try:
        assert len(row) == n_rows
    except:
        print("Must be a square matrix")
    try:
        assert math.fsum(row) == 1
    except:
        print("Rows must add up to 1")
```

```python
graph = Digraph('G')

for i, state_vector in enumerate(transition_matrix):
    origin_label = labels[i]
    graph.node(origin_label)
    for j, prob in enumerate(state_vector):
        destination_label = labels[j]
        graph.edge(origin_label,destination_label, label = str(prob))

graph
```
