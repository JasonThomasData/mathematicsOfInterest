### Competition via consumption of resources (Lotka Volterra Models)

These can be used to understand how two populations compete with each other:

- Via exploitation - one species affects another by consuming a shared resource. This is an indirect effect.
- Via interference - one species interferes with another's ability to use a resource.

This notebook concentrates on the first but there is a separate notebook for the second

### Species affect each other via exploitation

If left on its own, either species will approach the carrying capacity for the environment. When they compete there are 4 possible scenarios:

- $ N_1 $ causes the extinction of $ N_2 $
- $ N_2 $ causes the extinction of $ N_1 $
- $ N_1, N_2 $ are able to coexist
- $ N_1, N_2 $ are extinct

Consider a model that describes each population:

$ \frac{dN_1}{dt} = r_1 N_1(1 - \frac{N_1}{K_1}) $

$ \frac{dN_2}{dt} = r_2 N_2(1 - \frac{N_2}{K_2}) $

Perhaps this is two petri dishes containing each a different bacteria, and some substrate in each. Each dish has its own carrying capacity $ K $.

A way to think about the logistic models is that it displays the intra-species competition. That is, competition between the members of a species amongst themselves. That's why as the population approaches the carrying capacity, then the growth rate slows.

#### Lotka-Volterra Competition Model

Then to include a notion of inter-species competition, we can say:

$ \frac{dN_1}{dt} = r_1 N_1 \left(1 - \frac{N_1 + \alpha N_2}{K_1} \right) = r_1 N_1 \left(\frac{K_1 - N_1 - \alpha N_2}{K_1} \right) $

$ \frac{dN_2}{dt} = r_2 N_2 \left(1 - \frac{N_2 + \beta N_1}{K_2} \right) = r_2 N_2 \left(\frac{K_2 - N_2 - \beta N_1}{K_2} \right) $

This is now a system of non-linear differential equations.

To have a look at $ \alpha, \beta $ here, each is the relative importance of intra-species vs inter-species competition:

| $ \alpha, \beta $ | meaning                                                              |
| ----------------- | -------------------------------------------------------------------- |
| $ > 1 $           | inter-species competition is stronger than intra-species competition |
| $ < 1 $           | inter-species competition is weaker than intra-species competition   |

#### Equilibria points

##### Populations both extinct

If we set both differential equations to be $ 0 $.

$ 0 = r_1 N_1(1 - \frac{N_1 + \alpha N_2}{K_1}) $

$ 0 = r_2 N_2(1 - \frac{N_2 + \beta N_1}{K_2}) $

Then, $ N_1^*, N_2^* = 0 $. So if both populations are $ 0 $ then this point is an equilibrium. This is trivial.

##### One population lives and another goes extinct.

$ N_1^* = K_1, N_2^* = 0 $

This means that $ N_1 $ eventually gets to the carrying capacity as if just it was in the environment without $ N_2 $.

$ N_1^* = 0, N_2^* = K_2 $

This is the opposite case.

##### Populations coexist

I want to learn how to derive these:

$ N_1^* = \frac{K_1 - \alpha K_2}{1 - \alpha \beta} $

$ N_2^* = \frac{K_2 - \beta K_1}{1 - \alpha \beta} $

### Qualitative view

Below there is a phase plane of this system:

If the combination of $ N_1, N_2 $ falls on this line then the population of $ N_1 $ will not shrink or grow. If the combination falls below the nullcline. This line is $ \frac{dN_1}{dt} = 0 $

If the combination falls below the nullcline, in the yellow, then $ N_1 $ moves to the right. AKA $ \frac{dN_1}{dt} > 0 $

If the combination falls above the nullcline, in white, then $ N_1 $ moves to the left. AKA $ \frac{dN_1}{dt} < 0 $

Then the nullcline seems to represent the equilibrium for just this population.
