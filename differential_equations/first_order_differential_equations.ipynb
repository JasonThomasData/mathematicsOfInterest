### Solving differential equations

The strategy should be:

- Is the DE seperable?
    - Then solve it
- Is it a homogenous equation?
    - Try to make it seperable by substituting $ v = \frac{y}{x} $
    - Else solve it as a homogenous equation (linearity principle)
- Is the DE a linear DE, and non-homogenous?
    - Then solve it by using integrating factor
    - Else try the extended linearity principle
    - Can you apply the Bernoulli Differential Equation to solve it with the integrating factor

### Seperable DE

These can be arranged to have only one type of symbol on each side.

Eg:

Solve $ \frac{dy}{dx} = xy $, consider $ 3 = y(2) $

Can be separated like:

$ \frac{1}{y}\cdot dx = x\cdot dx $

So as long as you can get both sides with the same symbols in terms of $dy, dx$ then it's seperable. Then proceed to solve

$ \int \frac{1}{y}\cdot dx = \int x\cdot dx $

$ \ln y = \frac{x^2}{2} + C$

Be sure to remember the $ +C $ because it is often paramount.

$ y = e^{(x^2/2) + C}$

$ y = e^{(x^2/2)} e^C$

Since $ C $ is constant then $e^C$ will also be constant. Let:

$ A = e^C $

$ y = Ae^{(x^2/2)} $

This is the general solution but what about the initial condition?

$ 3 = Ae^{(2^2/2)} $

$ 3 = Ae^{2} $

$ A = \frac{3}{e^{2}} $

### Linear differential equation, is it homogeneous?

A linear equation is in the form:

$ \frac{dy}{dx} + P(x)y = Q(x) $

If $ Q(x) $ = 0 then the function

$ \frac{dy}{dx} + P(x)y = 0 $

Is homogenous.

#### Examples

$ \frac{dy}{dx} = sin(x)y $

This has no $ Q $ term so it is homogeneous.

$ \frac{dy}{dx} = sin(x) + y $

This has a $ Q $ term so it is non-homogeneous.


### Non-seperable linear non-homogeneous DE, solve using integrating factor $ I(x) $

When a DE is not seperable, if you can get it into the form:

$ \frac{dy}{dx} + P(x)y = Q(x) $

Then that is a `first order linear differential equation`. It's first order because the derivative is the first derivative.

Then it can be solved using the `Integrating Factor`.

This must be a non-homogeneous differential equation.

The integrating factor is always:

$ I(x) = e^{\int P(x) dx} $

To solve the DE, we should use the formula:

$ y = \frac{1}{I(x)} \int I(x) Q(x) dx $

#### Deriving the integrating factor from any linear DE

Consider the DE:

$ x\frac{dy}{dx} + y = x \cos(x) $

$ \frac{dy}{dx} + \frac{1}{x} y = \cos(x) $

Then this fits the form of a linear DE, where:

$ P(x) = \frac{1}{x} $

$ Q(x) = \cos(x) $

Now it's time to introduce an integrating factor. Let it be `I(x)`:

$ I(x)\frac{dy}{dx} + I(x)P(x)y = I(x)Q(x) $

Let's rearrange that:

$ I(x)Q(x) = I(x)\frac{dy}{dx} + I(x)P(x)y $

This almost looks like the product rule of differentiation, so let's make it look more like the product rule:

$ \frac{d}{dx}[I(x)Q(x)] = I(x)\frac{dy}{dx} + I(x)P(x)y $

But that would imply that:

$ \frac{dI}{dx} = I(x)P(x) $

That's progress, because it's a seperable DE.

$ \frac{1}{I(x)} \frac{dI}{dx} = P(x) $

$ \frac{1}{I(x)} dI = P(x) dx $

$ \int \left(\frac{1}{I(x)}\right) dI = \int(P(x))dx $

$ \ln(I(x)) = \int(P(x))dx $

$ I(x) = e^{\int(P(x))dx} $

Note, what's above is the integrating factor for any linear DE in the standard form.

$ I(x) = e^{\int(1/x)dx} $

$ I(x) = e^{\ln(x) + C} $

Since $ C $ is a constant then so is $ e^C $. 

Let $ A = e^C $

$ I(x) = Ae^{\ln(x)} $

$ I(x) = Ax $

Then let's revisit where we applied the integrating factor everywhere:

$ Ax\frac{dy}{dx} + AxP(x)y = AxQ(x) $

The constant, $ A $, cancels. So too could $ x $ but we need it to remain.

$ x\frac{dy}{dx} + xP(x)y = xQ(x) $

$ x\frac{dy}{dx} + x\frac{1}{x}y = x\cos(x) $

$ x\frac{dy}{dx} + 1\cdot y = x\cos(x) $

Above we don't strictly need the 1 to be there, but it's the to demonstrate that we have completed the product rule. The $ x $ on the far left has a derivative of $ 1 $.

Then this implies that:

$ \frac{d}{dx}[x\cdot y] = x\frac{dy}{dx} + 1 \cdot y = x\cos(x) $

But actually we are done with that middle part, so:

$ \frac{d}{dx}[x\cdot y] = x\cos(x) $

$ \int \left(\frac{d}{dx}[x\cdot y]\right)dx = \int(x\cos(x))dx $

$ x\cdot y = \int(x\cos(x))dx $

Now we can solve the RHS with integration by parts.

Let:

$ u = x $

$ v' = \cos(x) $

$ \int(x \cos(x))dx = x \sin(x) - \int(1 \sin(x)) dx $

$ \int(x \cos(x))dx = x \sin(x) + \cos(x) + C $

Now we can substitute the RHS of the previous equation:

$ x \cdot y = x \sin(x) + \cos(x) + C $

$ y = \sin(x) \frac{\cos(x) + C}{x} $
