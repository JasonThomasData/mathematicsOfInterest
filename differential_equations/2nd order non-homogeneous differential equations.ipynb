#### Motivation

Many real world systems take the form of 2nd order DEs. These are often harmonic oscillators and there is a notebook about this application.

### 2nd order non-homogeneous differential equations

The key here is these are non-homogeneous. See the other notebook about homogeneous equations.

This is an example that we would get from thinking about Harmonic Oscillators.

$ m \frac{d^2y}{dt^2} + b \frac{dy}{dt} + ky = f(t) $

If $ m \ne 0 $ then we can change the variables by dividing $ m $

$ \frac{d^2y}{dt^2} + p \frac{dy}{dt} + qy = g(t) $

It would be relatively easy to find the homogeneous equation associated with this:

$ \frac{d^2y}{dt^2} + p \frac{dy}{dt} + qy = 0 $

So let's do that to make our life easier.

Note that on the phase plane then this homogeneous DE is a stable sink while $ p, k > 0 $.

#### Using the extended linearity principle

There is a notebook on this method.

Suppose that:

$ y_h = c_1 y_1(t) + c_2 y_2(t) $ is a general solution of the homogeneous equation. Recall above it was mentioned this is a stable sink and it tends towards $ 0 $.

$ y_p $ is a particular solution to the non-homogeneous equation

$ y = c_1 y_1(t) + c_2 y_2(t) + y_p $ is the general solution for the non-homogeneous equation

$ y = y_h + y_p $

Since $ y_h $ tends to $ 0 $ then:

$ y \approx y_p $ as $ t \to \infty $

But my observation would be that $ y_p $ (the non-homogeneous particular solution) tends towards its equilibrium while the $ y_h $ (homogeous particular solution) tends towards its equilibrium of $ 0 $. So then these two combined means the system at large tends towards a combination of the two.

In the lecturer's words, the natural response dies while the forcing responce remains (in terms of harmonic oscillators).

#### A general method for finding the general solution

Given:

$ m \frac{d^2y}{dt^2} + b \frac{dy}{dt} + ky = f(t) $

$ \frac{d^2y}{dt^2} + p \frac{dy}{dt} + qy = g(t) $

- Find $ y_h $, which is the general solution to the homogenous equation

- find $ y_p $, which is the particular solution to the larger homnogeneous equation. Do this using the method of undetermined coefficients (which involves guessing). The notebook on the Linearity Principle includes tips for guessing.

- The general solution will be $ y_g = y_h + y_p $

### Example

Consider:

$ \frac{d^2y}{dt^2} + 5\frac{dy}{dt} + 6y = e^{-t} $

Let the homogeneous part be:

$ \frac{d^2y}{dt^2} + 5\frac{dy}{dt} + 6y = 0 $

Then for this the eigenvalues are: $ \lambda_1 = -2, \lambda = -3 $

And this has the general solution:

$ y_h = c_1 e^{-2t} + c_2 e^{-3t} $

Now we need a particular solution for the non-homogeneous equation:

$ \frac{d^2y}{dt^2} + 5\frac{dy}{dt} + 6y = e^{-t} $

And we can guess that $ y = ke^{-t} $

$ ke^{-t} - 5ke^{-t} + 6ke^{-t} = e^{-t} $

$ ke^{-t} - 5ke^{-t} + 6ke^{-t} = e^{-t} $

$ 2ke^{-t} = e^{-t} $

Then this is true if $ k = \frac{1}{2} $

$ y_p = \frac{1}{2}e^{-t} $

Then using the extended linearity principle, the general solution to the non-homogeneous equation is:

$ y_g = c_1 e^{-2t} + c_2 e^{-3t} + \frac{1}{2}e^{-t}$

### Example with Trig function RHS

Consider: 

$ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 6y = \cos t $

Now find the solution for the homogeneous differential equation like normal:

$ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 6y = 0 $

Then these have complex roots: $ -2 \pm i\sqrt{2} $

Then $ y = e^{(-2 + i\sqrt{2})t} $

$ e^{(-2 + i\sqrt{2})t} = e^{-2t} e^{(\sqrt{2})it} $

We employ Euler's Identity:

$ e^{-2t} e^{(\sqrt{2})it} = e^{-2t} \left(\cos (\sqrt{2})t + i \sin (\sqrt{2})t \right) $

With the theorem on page 304 of Blanchard, we can now say that the general solution is this:

$ y_h = c_1 e^{-2t} \cos (\sqrt{2})t + c_2 e^{-2t} \sin (\sqrt{2})t $

Now to find the particular solution $ y_p $

Guess:

$ y = \alpha \cos t + \beta \sin t $

$ (- \alpha \cos t - \beta \sin t ) + 4(- \alpha \sin t + \beta \cos t ) + 6(\alpha \cos t + \beta \sin t ) = \cos t $

$ (- \alpha \cos t - \beta \sin t ) + (- 4\alpha \sin t + 4\beta \cos t ) + (6\alpha \cos t + 6\beta \sin t ) = \cos t $

$ 4\beta \cos t + 5\alpha \cos t + -4\alpha \sin t + 5\beta \sin t = \cos t $

$ (\cos t )(5\alpha + 4\beta) + (\sin t)(-4\alpha + 5\beta) = \cos t $

Then we can use an augmented matrix to solve the system:

Note the first line is $ 1 $ because it is a multiple of $ \cos t $, as is the RHS.

```octave
format rat;
M = [5 4 1; -4 5 0]

M(1,:) = M(1,:) + M(2,:);
M(2,:) = M(2,:) + M(1,:)*4;
M(2,:) = M(2,:)*(1/41);

M(1,:) = M(1,:) + M(2,:)*(-9)
```

Then these are the values of $ \alpha, \beta $

This implies:

$ y_p = \frac{5}{41} \cos t + \frac{4}{41} \sin t $

And finally this is the general solution for the non-homogeneous equation:

$ y_g = c_1 e^{-2t} \cos (\sqrt{2})t + c_2 e^{-2t} \sin (\sqrt{2})t + \frac{5}{41} \cos t + \frac{4}{41} \sin t $

### Example with complex RHS

Again, consider:

$ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 6y = \cos t $

But let's replace it with:

$ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 6y = e^{it} $

Then thanks to Euler's formula:

$ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 6y = \cos t + i \sin t $

Then we substitute $ y_c = y_{re} + i y_{im} $:

$ \frac{d^2(y_{re} + i y_{im})}{dt^2} + 4\frac{d(y_{re} + i y_{im})}{dt} + 6(y_{re} + i y_{im}) = \cos t + i \sin t $

Then what we have derived is a combination of real and imaginary parts:

$ \frac{d^2y_{re}}{dt^2} + 4\frac{dy_{re}}{dt} + 6y_{re}) = \cos t $

$ \frac{d^2 i y_{im}}{dt^2} + 4\frac{d i y_{im}}{dt} + 6 i y_{im} = i \sin t $

Then what this means is that given a trig function on the RHS, you can replace it with $ e^{it} $ and then find the real part as above, then you could find the $ y_p $ this way, by dealing with complex numbers.

### Example without the first order derivative

Th e common real-world example is a harmonic oscillator with no dampening but also it has a forcing part.

$ \frac{d^2y}{dt^2} + q^2 y = \cos \omega t $

Using the method from earlier where we switched the RHS for a complex statement:

$ \frac{d^2y}{dt^2} + q^2 y = e^{i \omega t} $

The easy part is finding the homogeneous solution.

The eigenvalues are: $ \pm iq $

$ y_h = c_1 \cos qt + c_2 \sin qt $

By using Euler's Identity we can convert it to:

$ \frac{d^2y}{dt^2} + q^2 y = ( \cos (\omega t) + i \sin (\omega t)) $

Now, choose the solution $ y_c = ae^{i \omega t} $

$ \frac{d^2 ae^{i \omega t}}{dt} + q^2 ae^{i \omega t} = ( \cos (\omega t) + i \sin (\omega t)) $

$ a i^2 \omega^2 e^{i \omega t} + q^2 ae^{i \omega t} = ( \cos (\omega t) + i \sin (\omega t)) $

$ -\omega^2 ae^{i \omega t} + q^2 ae^{i \omega t} = ( \cos (\omega t) + i \sin (\omega t)) $

$ (q^2 - \omega^2)ae^{i \omega t} = ( \cos (\omega t) + i \sin (\omega t)) $

There are two equivalent equations:

$ \frac{d^2y}{dt^2} +  q^2 y = e^{i\omega t} $

$ \frac{d^2y_c}{dt^2} +  q^2 y_c = a(q^2-\omega^2) e^{i\omega t} $

So therefore:

$ e^{i \omega t} = a(q^2-\omega^2) e^{i\omega t} $

$ 1 = a(q^2-\omega^2) $

This implies two results:

(1)

$ e^{i \omega t} = ( \cos (\omega t) + i \sin (\omega t)) $

(2)

$ a = \frac{1}{q^2-\omega^2} $

Then substitute this back into the guessed solution $ y_c = ae^{i \omega t} $

$ y_c = \frac{1}{q^2 - \omega^2} ( \cos (\omega t) + i \sin (\omega t)) $

This, above, is the complex solution. We need to isolate just the real part to find the particular solution to the non-homogeneous equation:

$ y_p = \frac{1}{{q^2 - \omega^2}}\cos (\omega t) $

And then you combine it with $ y_h $

$ y_g = c_1 \cos qt + c_2 \sin qt + \frac{1}{{q^2 - \omega^2}}\cos (\omega t) $

_These formulas only apply when $ \omega \ne q $, otherwise you end up with division by 0_

#### Same example, with initial coniditions

Now suppose we want to find a particular solution to the non-homogeneous differential equation, by using the general solution. That is, we don't want to use the particular solution that we used to find the general solution because we want to determine the values of $ c_1, c_2 $


Given the initial conditions:

$ y_g(0) = 0 $

$ y_g'(0) = 0 $

And given the general solution and also its derivative:

$ y_g = c_1 \cos qt + c_2 \sin qt + \frac{1}{{q^2 - \omega^2}}\cos (\omega t) $

$ y_g' = q c_1 (-\sin qt) + q c_2 \cos qt + \frac{1}{{q^2 - \omega^2}} (\omega -\sin (\omega t)) $

Substituting in the initial condition for the general solution:

$ y_g(0) = 0 = c_1 \cos (0) + c_2 \sin (0) + \frac{1}{{q^2 - \omega^2}}\cos (0) $

$ 0 = c_1 (1) + c_2 (0) + \frac{1}{{q^2 - \omega^2}}(1) $

$ 0 = c_1 + \frac{1}{{q^2 - \omega^2}} $

$ c_1 = - \frac{1}{{q^2 - \omega^2}} = -a $

Substituting in the initial condition for the general solution's derivative:

$ y_g'(0) = 0 = g c_1 (-\sin q(0)) + q c_2 \cos q(0) + \frac{1}{{q^2 - \omega^2}} (\omega -\sin (\omega (0))) $

$ 0 = q c_2 + \frac{1}{{q^2 - \omega^2}} (0) $

$ c_2 = 0 $

So finally we can substitute in these constants to find:

$ y_g = (-\frac{1}{{q^2 - \omega^2}}) \cos qt + \frac{1}{{q^2 - \omega^2}}\cos (\omega t) $

$ y_g = -\frac{\cos qt}{{q^2 - \omega^2}}  + \frac{\cos \omega t}{{q^2 - \omega^2}} $

Recall that:

$ a = \frac{1}{q^2 - \omega^2} $

So the particular solution obtained by using $ c_1, c_2 $ is:

$ y = a(\cos \omega t - \cos qt) $

### Hints for guessing the $ y_p $ part in the process

Given:

$ \frac{d^2y}{dt^2} + p \frac{dy}{dt} + qy = g(t) $

Then you need to guess some $ y_p $

- $ g(t) $ is the product of a poynomial and an exponential then try $ y_p $ as the same product function where the polynomial has the same degree.
