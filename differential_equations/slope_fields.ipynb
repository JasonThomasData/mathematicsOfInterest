### Slope fields (qualitative analysis)

A slope field is a method for visualising the solutions for differential equations. This slope field is represented as a vector field. The slopes are actually vectors and the slopes are not on the ticks (have a look at it).

Given a first order differential equation, we can use a slope field to see what the solution would be, depending on a given initial condition.

Without finding a solution you can understand the overall behaviour of solutions.

The function $ \sin(x^2) $ is known to have no simple integral formula. We can use a slope field to see how it behaves with respect to $ x $.

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = sin(x.^2);
end

slope_field(@fdash, [-3 3 -3 3])
```

The function $ x^2 $ is elementary and its family of solutions are indeed simple.

```octave
clear 

function dy_dx = fdash(x, y)
    dy_dx = x.^2;
end

slope_field(@fdash, [-5 5 -5 5])
```

What if we have an initial condition? Then we can use that, and the general solution, to find a particular solution.

$ \frac{dy}{dx} = x^2 $

$ y = \frac{x^3}{3} + c $, which is the family of solutions.

Let $ (x,y) = (2,1) $ be the initial condition.

$ 1 = \frac{2^3}{3} + c $

Then solve for $ c $ to find the particular solution.

$ c = 1 - \frac{2^3}{3} $

$ c = 1 - \frac{8}{3} $

$ c = -\frac{5}{3} $

Then the particular solution is:

$ y = \frac{x^3}{3} - \frac{5}{3} $, given the initial condition of $ (2,1) $




```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = x.^2;
end

# You would typically define the initial condition before finding the particular solution's + c

initial_condition = [2 1];

function solution = particular_solution(x, y)
    solution = (x.^3)/3 - (5/3);
end

slope_field(@fdash, [-5 5 -5 5], @particular_solution, initial_condition);
```

You might have noticed that given an $ x $ value, that all of the $ y $ values are parallel. That is the case if:

$ \frac{dy}{dx} = f(x) $

But not when:

$ \frac{dy}{dx} = f(x, y) $

Or when the equation is autonomous:

$ \frac{dy}{dx} = f(y) $

Let:

$ \frac{dy}{dx} = y x^2 $

Then notice that the lines are not parallel along any tick value:

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = y.* x.^2;
end

slope_field(@fdash, [-5 5 -5 5])
```

The equation is integrable because it is seperable.

$ \frac{1}{y} dy = x^2 dx $

$ \int \frac{1}{y} dy = \int x^2 dx $

$ \ln y = \frac{x^3}{3} + c_1 $

$ y = e^{\frac{x^3}{3} + c_1} $

$ y = e^{\frac{x^3}{3}}e^{c_1} $

A family of solutions:

$ y = c_2 e^{\frac{x^3}{3}} $

If the initial condition is $ (x, y) = (1, 2) $, then:

$ (2) = c_2 e^{\frac{(1)^3}{3}} $

$ c_2 = \frac{2}{\sqrt[3]{e}} $

This means that given $ (x,y) = (1,2) $ then the particular solution is:

$ y = \left( \frac{2}{\sqrt[3]{e}} \right) e^{\frac{x.^3}{3}} $


```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = y.* x.^2;
end

initial_condition = [1 2];

function solution = particular_solution(x, y)
    solution = (2/cbrt(e)) .* exp(x.^3/3);
end

slope_field(@fdash, [-5 5 -5 5], @particular_solution, initial_condition);
```
