#### Extended linearity principle for non-homogeneous DE

The general solution for a non-homogenous equation is:

For some multiple $ k $

$ y_g(x) = y_p(x) + ky_h(x) $

And let the subscripts be:

$ g $ is the general solution

$ p $ is the particular solution

$ h $ is the homogeneous solution

Which means, if you have a non-homogeneous linear differential equation, you should solve the homogeneous side first, and then add the solution as a multiple.

$ \frac{dy}{dx} = cos(x)y + \frac{1}{5}(1 - x cos(x)) $

This one looks hard to solve, but we can solve the homogeneous equation with the rest removed, first.

We saw from earlier that the solution to the homogeneous part is:

$ y_h = k e^{-sin(x)} $

Then for this entire non-homogeneous part, we can say:

$ y_g = y_p + k e^{-sin(x)} $

Finding $ y_p $ is often guesswork.

Looking at the non-homogeneous equation again:

$ \frac{dy}{dx} = cos(x)y + \frac{1}{5}(1 - x cos(x)) $

Let's suppose the a solution to the equation is:

$ y = \alpha x, \frac{dy}{dx} = \alpha $

Then:

$ \frac{dy}{dx} = \cos(x) \alpha x + \frac{1}{5}(1 - x \cos(x)) $

$ \frac{dy}{dx} = \cos(x) \alpha x + \frac{1}{5} - \frac{1}{5} x \cos(x) $

$ \frac{dy}{dx} = \cos(x) \alpha x - \frac{1}{5} x \cos(x) + \frac{1}{5} $

$ \frac{dy}{dx} = \cos(x) (\alpha x - \frac{1}{5}x) + \frac{1}{5} $

Then if $ \alpha = 1/5 $, then 

$ \frac{dy}{dx} = \cos(x) (\frac{1}{5} x - \frac{1}{5}x) + \frac{1}{5} $

$ \frac{dy}{dx} = \frac{1}{5} $

Then this is a particular solution to the non-homogeneous equation.

The general solution is:

$ y_g(x) = y_p(x) + ky_h(x) $

$ y_g(x) = \frac{1}{5} + k e^{-sin(x)} $

#### Guessing a particular solution

Given a non-homogeneous linear differential equation, then guessing the $ y $ for:

$ \frac{dy}{dt} + P(x)y = Q(x) $

Then use the guess to find the value of $ \alpha $ that was introduced below:

Tips for guessing $ y $:

- If $Q(x) $ is a poynomial, guess to the same degree
- If $Q(x) = Ce^{mt} $ then try: $\alpha e^{mt} $ and if it fails try $ \alpha te^{mt} $
    - A reason the first case above might fail is if the guess appears in the homogeneous equation. Then try the second case above
- If $Q(x) = \sin(mt) $ then try: $\alpha \sin(mt) + \beta \cos(mt) $ and if it fails try $ \alpha t\sin(mt) + \beta t\cos(mt) $
- If $Q(x) = \cos(mt) $ then try: $\alpha \cos(mt) + \beta \sin(mt) $ and if it fails try $ \alpha t\cos(mt) + \beta t\sin(mt) $
