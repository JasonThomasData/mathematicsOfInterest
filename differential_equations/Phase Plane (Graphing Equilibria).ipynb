### Graphing Equlibria

If you have any model that has equilibria, then you might like to know if it has stable or unstable equilibria.

Consider the logistic population mode. It has two equilibria.


```octave
clear

function dP_dt = fdash(t, P)
    r = 1;
    K = 3;
    dP_dt = r.*P.*(1.-(P/K));
end

initial_condition = [0,0.05];

plot_dimensions = [-11 11 -11 11];

function solution = particular_solution(t, P)
    r = 1;
    K = 3;
    P_0 = 0.05;
    solution = K./(1 .+ (K.-P_0)./P_0 .* exp(-r.*t));
end
slope_field(@fdash, plot_dimensions, @particular_solution, initial_condition);
```

To find if they are stable or unstable, we can use a graph to show where the function has a positive or negative rate of growth. The real roots are the equilibria.

```octave
clear;

t = linspace(-10,10);

r = 1;
K = 3;
P_0 = 0.05;

P = K./(1 .+ (K.-P_0)./P_0 .* exp(-r.*t));
dP_dt = r.*P.*(1.-(P/K));

plot(P, dP_dt);
line([-1 6], [0 0]);
axis([-1 6 -1 6]);
axis square;
```

Since the equilibria are the real roots, then when $ P ~ 0 $ the function grows. When $ P ~ K/2 $ it continues growing, but at a more stable pace. When $ P = K $ then the growth stops. Then $ P = K $ is a stable equilibrium. 

I think the Theta-Logistic model would have a different shape but would have the same real roots.

### Allee effect

Take the modified logistic model and see this again:

```octave
clear;

t = linspace(-10,10);

r = 1;
K = 5;
P_0 = 0.05;
M = 2;

P = K./(1 .+ (K.-P_0)./P_0 .* exp(-r.*t));
dP_dt = r.*P.*(1.-(P./K)).*(P./M-1);

plot(P, dP_dt);
line([-1 6], [0 0]);
axis([-1 6 -1 6]);
axis square;
    
```

If the population drops below the critical threshold of $ M $ then it will experience negative growth. Below 0 and this decreases to $ 0 $, which is a stable equilibrium. $ P = 2 $ here is an unstable equilibrium and either side of it the population will trend towards other values.
