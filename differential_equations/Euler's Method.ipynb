### Euler's Method

This is a basic method of approximation, otherwise called a numerical method.

If you have some differential equation. Let's pretend that for whatever reason, we don't know the solution. For this we can use a differential equation that's easily seperable so we can plot the numerical solution against the analytic solution:

$ \frac{dy}{dx} = y $

$ \frac{1}{y} dy = dx $

$ \ln(y) = x + c_1 $

$ y = e^{x + c_1} $

Then this is the analytic solution, and is a general solution:

$ y = c_2 e^{x} $

Let's plot a particular solution on a slope field:

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = y;
end

initial_condition = [0,1];

function solution = unknown_particular_solution(x, y)
    solution = 1 * exp(x);
end

plot_dimensions = [-5 5 -5 5];

slope_field(@fdash, plot_dimensions, @unknown_particular_solution, initial_condition);
```

Below, we can use Euler's Method to approximate this.

Recall:

$ \frac{dy}{dx} = y $

The differential equation tells us that at any given point of $ x $, the rate of change is equal to $ y $.

So, at the point $ (0,1) $ then the rate of change is $ 1 $.

Then the change to $ x = 1 $ is the previous $ y $ plus the change.

| x | y   | y'  |
|---|-----|-----|
| 0 | 1   | 1   |
| 1 | 2   | 2   |
| 2 | 4   | 4   |
| 3 | 8   | 8   |

This is an approximation and let's see it graphed:

```octave
plot_dimensions = [-5 5 -5 5];

function plot_approx(domain_to_approximate, plot_dimensions, differential_equation, initial_y)
    approximation = euler_method(differential_equation, domain_to_approximate, initial_y);
    plot(domain_to_approximate, approximation);
    axis(plot_dimensions);
    axis square;
end

plot_approx(linspace(0,2,3), plot_dimensions, @fdash, 1);
```

```octave
plot_approx(linspace(0,2,4), plot_dimensions, @fdash, 1);
```

```octave
plot_approx(linspace(0,2,7), plot_dimensions, @fdash, 1);
```

What about for a function that is hard to solve analytically?

$ \frac{dy}{dx} = sin(x^2) $

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = sin(x.^2);
end

initial_condition = [0,0];

plot_dimensions = [-5 5 -5 5];

slope_field(@fdash, plot_dimensions);

hold on;

plot_approx(linspace(-5,5,30), plot_dimensions, @fdash, 0);
```
