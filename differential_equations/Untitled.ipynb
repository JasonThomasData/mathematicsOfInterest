### Euler's Method

This is a basic method of approximation, otherwise called a numerical method.

If you have some differential equation. Let's pretend that for whatever reason, we don't know the solution.  But of course it's easily seperable:

$ \frac{dy}{dx} = y $

$ \frac{1}{y} dy = dx $

$ \ln(y) = x + c_1 $

$ y = e^{x + c_1} $

$ y = c_2 e^{x} $

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = y;
end

initial_condition = [0,1];

function solution = unknown_particular_solution(x, y)
    solution = 1 * exp(x);
end

slope_field(@fdash, [-5 5 -5 5], @unknown_particular_solution, initial_condition);
```

The differential equation tells us that at any given point of $ x $, the rate of change is equal to $ y $.

So, at the point $ (0,1) $ then the rate of change is $ 1 $.

Then the change to $ x = 1 $ is the previous $ y $ plus the change.

| x | y   | y'  |
|---|-----|-----|
| 0 | 1   | 1   |
| 1 | 2   | 2   |
| 2 | 4   | 4   |
| 3 | 8   | 8   |

This is an approximation and let's see it:

```octave
clear

function dy_dx = fdash(x, y)
    dy_dx = y;
end

initial_condition = [0,1];

function solution = unknown_particular_solution(x, y)
    solution = 1 * exp(x);
end

slope_field(@fdash, [0 8 0 8], @unknown_particular_solution, initial_condition);

hold on;

plot([0,1,2,3],[1,2,4,8]);
```

This optimisation would be better if it had more steps with smaller gaps in between. This process is implemented in the function _euler_method.m_
