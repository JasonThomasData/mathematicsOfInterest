### Harmonic oscillators

These can be modelled using second order DEs.

Harmonic oscillators usually, and certainly in these notebooks, have the following interpretations:

$ m $ is mass

$ b $ is friction 

$ k $ is the restoring force, eg spring strength

Here are some definitions:

- Undamped, unforced $ m \frac{d^2}{dt^2} + b \frac{dy}{dt} + ky = 0 $

- Damped, unforced $ m \frac{d^2}{dt^2} + ky = 0 $

- Undamped, forced $ m \frac{d^2}{dt^2} + b \frac{dy}{dt} + ky = f(t) $

- Damped, forced $ m \frac{d^2}{dt^2} + ky = f(t) $

#### Examples

The pendulum and the spring-mass systems are both simple harmonic oscilators, and that means each will experience acceleration to its equilibrium. For the spring that's the rest position of the mass. For the pendulum that's the pendulum pointing downwards. They're simple in the sense that there's no damping effects and 

If we assume both systems have no dampening effects then that makes them Simple Harmonic Oscillators.

#### Regarding springs

![spring-mass](springmass.png)

Say we have some spring with a load on the end. The function $ y(t) $ is the distance away from the rest position.

Then another word to describe $ y $ is "displacement"

Then speed is:

$ \frac{dy}{dt} $

And acceleration is:

$ \frac{d^2y}{dt^2} $

Acceleration is used in Newton's second law:

$ F = ma $

There is another equation that defines force, and it is Hooke's Law:

$ F = -ky $

Here $ y $ is again displacement and $ k $ is the restoring force of the spring. The stronger the spring, the more force the spring exerts.

Since Newton's Second Law and Hooke's Law define $ F $ then:

$ F = m \frac{d^2y}{dt^2} = -ky $

Then the homogeneous equation we are asked to derive is:

$ m\frac{d^2y}{dt^2} + ky = 0 $

#### Regarding pendulums

We can imagine a pendulum hangs towards the Earth's centre of gravity.

The angle at rest is $ \theta = 0 $.

Suppose the pendulum has a fixed length and there a circle that has a radius of the same length.

If you increase $ \theta $ then you get something like this:

![pendulum1](pendulum1.png)

We should note that gravity is pulling the pendulum downward, but because the string/rod is assumed to be fixed then the pendulum moves at a tangent to the circle. I wrote $ m $ but it makes more sense as $ mg $, as in mass by gravitational constant.

![pendulum2](pendulum2.png)

The angle between the tangent line and the downwards pull of gravity is a right angle minus $theta$, or $ \frac{pi}{2} - \theta $

![pendulum3](pendulum3.png)

We can draw one more line and to construct it we require that it has a right angle to the tangent line

![pendulum4](pendulum4.png)

This new triangle has a right angle. Suppose we know the force of gravity (hypotenuse) and we know the angle, then we can use this fact:

$ \sin \theta = \frac{\text{opp.}}{\text{hyp.}} $

$ \text{hyp.} * \sin \theta = \text{opp.} $

Altogether we find the quantity of this tangent force is:

$ F = - mg \sin \theta $

The minus sign is included because the force is always acting towards the system equilibrium.

We need to include another definition of force.

If we imagine a unit circle then it has an inner angle of $ 2\pi $. Given the radius then we can find its circumference $ 2\pi r $. We can find an arc length of we know the angle $ \theta $ that covers the arc.

$ s = \theta r $

But it will look more sensible to call the radius $ l $ since it's a pendulum.

$ s = \theta l $

And since it's a pendulum then the arc length $ s $ is displacement. So therefore the second derivative is acceleration.

Newton's second law can be written in terms of the arc length $ s $:

$ F = m \frac{d^2 s}{dt^2} $

$ l $ is constant so the rate of change for acceleration entirely depends on theta:

$ \frac{d^2 s}{dt^2} = l \frac{d^2 \theta}{dt^2} $

We can now substitute it into the equation for Newton's Second Law:

$ F = m l \frac{d^2 \theta}{dt^2} $

And now we can combine the two force equations:

$ F = -mg \sin \theta = m l \frac{d^2 \theta}{dt^2} $

Do some cancelling:

$ -g \sin \theta =  l \frac{d^2 \theta}{dt^2} $

And finally:

$ \frac{d^2 \theta}{dt^2} = -\frac{g \sin \theta}{l} $

We now have a second order differential equation and we should be able to approximate it with the linearisation:

$ \sin x \approx x $

### Spring system with damping no forcing

Recall:

Newton's second law:

$ f = ma $

Hooke's Law:

$ f = -ky $

But damping reduces the effect of Hooke's Law,

$ f = -ky + b\frac{dy}{dt} $

Where $ b $ is the friction coefficient. Friction is proportional to velocity, so this has the first derivative included.

The put it all together:

$ ma = -ky + b\frac{dy}{dt} $

$ m\frac{d^2y}{dt^2} = -ky + b\frac{dy}{dt} $

$ m\frac{d^2y}{dt^2} - b\frac{dy}{dt} +ky $

Then this is a second order DE, so then we know that this is equivalent by solving for $ y $

$ m \lambda^2 - b \lambda + k $

Then the roots for this polynomial are:

$ \frac{b \pm \sqrt{-b^2 - 4mk}}{2m} $

Then this is enough to see that in the tau-delta plane, $ \tau=b $ is the trace and $ \Delta=mk $ are delta.

$ \Delta $ will always be positive, otherwise either the weight has no mass or the spring has no power.

If $ b $ is negative then you will see the system is a stable sink but if there is no friction then $ b = 0 $, meaning it sits on the centre, so it's neither a source nor a sink.
