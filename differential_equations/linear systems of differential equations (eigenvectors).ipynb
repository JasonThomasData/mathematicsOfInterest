### Linear systems of differential equations

- Analytic solutions are possible
- Many real-world phenomena are like this (eg predator-prey systems)
- Give us an understanding of non-linear systems, sometimes

Given two autonomous linear differential equations:

$ \frac{dx}{dt} = ax + by $

$ \frac{dy}{dt} = cx + dy $

They are autonomous in the sense that the independent variable $ t $ is not a part of either equation, although $ t $ is still the parameter of $ x, y $

$ a, b, c, d $ are constant with respect to $ t $, and are coefficients of $ x, y $

Let's represent this as:

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} ax + by \\ cx + dy \end{matrix} \right] $

We should recognise that this is the expanded for of matrix multiplication:

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} a, b \\ c, d \end{matrix} \right] \left[ \begin{matrix} x \\ y \end{matrix} \right]$

The vector of derivatives is just the derivative of a vector's components, so it's $ \dot X $. The coefficient matrix is $ A $ and the column matrix on the far right is $ \vec{X} $

$ \dot X = A \vec{X} $

Let's make an assumption about $ \vec{X} $. Let's suppose it contains the exponential function.

$ \vec{X(t)} = \vec{V} e^{\lambda t} $, where $ \vec{V} $ is a vector of constant values.

If we differentiate the above then:

$ \dot X = \lambda \vec{V} e^{\lambda t} $

But if we substitute $ X(t) $ into $ \dot X = A\vec{X} $ then:

$ \dot X(t) = A \vec{V} e^{\lambda t} $

And therefore:

$ A \vec{V} e^{\lambda t} = \lambda \vec{V} e^{\lambda t} $

$ A \vec{V} = \lambda \vec{V} $

We should recognise this as the form of the eigenvector problem.

There is a notebook in `LinearAlgebra/` explaining about Eigenvalues and Eigenvectors, and how to solve them using the notion of determinants.

#### Equilibrium points

Given the above example of creating a matrix from the system of equations (coefficients) then we can make this generalisation:

For any coefficient matrix $ A $, if $ \det(A) \ne 0 $ then the system has one equilibrium point at $ (0,0) $

#### Example for solving a linear system of differential equations

$ \frac{dx}{dt} = x + y $

$ \frac{dy}{dt} = 4x - 2y $

Then in matrix form:

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} 1, & 1 \\ 4, & -2 \end{matrix} \right] \left[ \begin{matrix} x \\ y \end{matrix} \right]$

$ \dot X = A \vec{X} $

Then by the work from above:

$ A \vec{V} = \lambda \vec{V} $

And by using the characteristic equation:

$ det(A - \lambda \vec{I}) = \vec{0} $

$ det \left( \left[ \begin{matrix} 1 - \lambda & 1 \\ 4 & -2 - \lambda \end{matrix} \right] \right) = \left[ \begin{matrix} 0 \\ 0 \end{matrix} \right] $

$ (1 - \lambda)(-2 - \lambda) - (1)(4) = 0 $

$ (\lambda^2 + 2\lambda - \lambda - 2) - 4 = 0$ 

$ \lambda^2 + \lambda - 6 = 0$

Then this is just a quadratic and:

$ (\lambda + 3)(\lambda - 2) = 0 $

Therefore the roots or Eigenvalues are: $ \lambda = -3, 2 $

We revisit this:

$ A\vec{V} = \lambda \vec{V} $

And using $ \lambda = -3 $

$ A\vec{V} = -3 \vec{V} $

$ B = A +3 \vec{I} = \vec{0} $

$ \left[ \begin{matrix} 4 & 1 \\ 4 & 1 \end{matrix} \right] = \left[ \begin{matrix} 0 \\ 0 \end{matrix} \right] $

$ V_1 = [-1, 4] = [x, y] $

And using $ \lambda = 2 $

$ A\vec{V} = 2 \vec{V} $

$ B = A -2 \vec{I} = \vec{0} $

$ \left[ \begin{matrix} -1 & 1 \\ 4 & -4 \end{matrix} \right] = \left[ \begin{matrix} 0 \\ 0 \end{matrix} \right] $

$ V_2 = [1, 1] = [x, y] $


```octave
### Verification of eigenvectors
[1 1; 4 -2]*[-1;4] == -3*[-1;4]
[1 1; 4 -2]*[1;1] == 2*[1;1]
```

```octave
x = linspace(-5, 5);

y_1 = -4 .*x; # [x,y]=[-1,4]
y_2 = x; # [x,y]=[1,1], just the identity function

p_1 = plot(x, y_1);
hold on;
p_2 = plot(x, y_2);

axis([-5 5 -5 5])
axis square;

line([-5 5], [0 0], "linestyle", ":", "color", "k")
line([0 0], [-5 5], "linestyle", ":", "color", "k")

legend([p_1 p_2], 'V_1 = [ -1, 4 ]', 'V_2 = [ 1, 1 ]');
```

### Straight line solutions and the general solution

If a matrix $ A $ has a real eigenvalue $ \lambda $ with associated eigenvector $ \vec{v} $.

The linear system $ \frac{dy}{dt} = AY $ has the straight line solution:

$ y = e^{\lambda t} \vec{v} $

If $ \lambda_1 \ne \lambda_2 $ and are real values then:

$ y_1 = e^{\lambda_1 t} \vec{v}_1 $

$ y_2 = e^{\lambda_2 t} \vec{v}_2 $

Then because of the linearity principle:

$ y = k_1 e^{\lambda_1 t} \vec{v}_1 + k_2 e^{\lambda_2 t} \vec{v}_2 $

And this is the general solution.

### Uncoupled system

If you have:

$ \frac{dx}{dt} = ax $

$ \frac{dy}{dt} = dy $

These are solveable on their own.

Despite this we can still find reasons to consider this as an uncoupled system.

- Solveable without linear algebra
- There are some qualitative solutions that can be used to deal with the uncoupled system

Then:

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} ax + 0y \\ 0x + dy \end{matrix} \right] $

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} a, 0 \\ 0, d \end{matrix} \right] \left[ \begin{matrix} x \\ y \end{matrix} \right]$

#### Example and phase plots (with real values, complex values have spiral plots)

$ a = -1, d = -1 $

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} -1x + 0y \\ 0x + -1y \end{matrix} \right] $

$ \left( \begin{matrix} x' \\ y' \end{matrix} \right) = \left[ \begin{matrix} -1, 0 \\ 0, -1 \end{matrix} \right] \left[ \begin{matrix} x \\ y \end{matrix} \right]$

Solve each equation by itself:

$ \frac{dx}{dt} = -x $

$ x = x_0 e^{-1t} $

$ \frac{dx}{dt} = -y $

$ y = y_0 e^{-1t} $

Now let's rearrange the first one:

$ e^{-1t} = \frac{x(t)}{x_0} $

And substitute into the second one:

$ y = y_0 \frac{x(t)}{x_0}$

$ y = x(t) \frac{y_0}{x_0}$

Notice in the phase plot below, that anything above 0 is decreasing towards 0, and anything below is going towards zero. So as $ t $ varies, the solutions will always tend along straight lines.

#### a = d

```octave
clear;

t = linspace(-100,100,1000);

a = -1;
d = -1;

#Initial conditions
ic_1 = [1 1];
ic_2 = [1 -1];
ic_3 = [-1 1];
ic_4 = [-1 -1];

# y = y_0 e^d
y_0 = ic_1(1)/exp(d);
# x = x_0 e^a
x_0 = ic_1(2)/exp(a);
y = y_0 * exp(d*t);
x = x_0 * exp(a*t);
plot(x, y);
hold on;

y_0 = ic_2(1)/exp(d);
x_0 = ic_2(2)/exp(a);
y = y_0 * exp(d*t);
x = x_0 * exp(a*t);
plot(x, y);

y_0 = ic_3(1)/exp(d);
x_0 = ic_3(2)/exp(a);
y = y_0 * exp(d*t);
x = x_0 * exp(a*t);
plot(x, y);

y_0 = ic_4(1)/exp(d);
x_0 = ic_4(2)/exp(a);
y = y_0 * exp(d*t);
x = x_0 * exp(a*t);
plot(x, y);

line([-2 2], [0 0]);
axis([-2 2 -2 2]);
axis square;
    
```

_Note: Above, all values move towards 0 as $ t \to \infty $, aka sink_
_Note: Above, if $ a = d $ but both were $ > 0 $ then this would be a source not a sink
