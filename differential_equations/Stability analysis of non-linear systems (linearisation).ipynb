### Stability analysis of non-linear systems (linearisation)

Consider a two-dimensional system of autonomous equations:

![system to study](Screenshot_2023-04-17_16-26-48.png)

And we can study the equilibrium of the system at $ (x^*, y^*) = (1, 1) $, which is the saddle in the middle. We can see that it's a saddle (qualitative) but can we also confirm it using stability analysis (analytic).

#### The analysis

The process is this:
- Change of variables: move the equilibrium to the origin
- Eliminate higher order (non-linear) terms
- Using the linearised system, find eigenvalues to do this

This is because once we have linearised the non-linear system via a change of variables, and when the equilbrium is at the origin, then eigenvalues will be useful ______.

Let:

$ u = x - 1, v = y - 1 $

This is a good choice because the system after changing variables will be shifted by this ammount, above. Then the new system will have its equilibrium at the origin $ (0,0) $.

$ \frac{du}{dt} = \frac{d(x-1)}{dt} = \frac{d(x)}{dt} $ 

Now rewrite $ u, v $:

$ x = u + 1, y = v + 1 $

And look at what we're working with.

$ \frac{dx}{dt} = 2x \left(1 - \frac{x}{2} \right) - xy $

Expand it:

$ \frac{dx}{dt} = \left(2x - x^2 \right) - xy $

Substitute in the change of variables.

$ \frac{du}{dt} = \left(2(u+1) - (u+1)^2 \right) - (u+1)(v+1) $

$ \frac{du}{dt} = \left((2u+2) - (u^2 + 2u + 1) \right) - (uv + u + v + 1) $

$ \frac{du}{dt} = 2u + 2 - u^2 - 2u - 1 - uv - u - v - 1 $

$ \frac{du}{dt} = - u^2 - uv - u - v $

Do the same for the other equation:

$ \frac{dy}{dt} = 3y \left(1 - \frac{y}{3} \right) - 2xy $

And do the same process as before, the result is:

$ \frac{dv}{dt} = -2u - v - 2uv - v^2 $ 

So now we have this system of equations with an equilibrium at $ (0,0) $

$ \frac{du}{dt} = - u^2 - uv - u - v $

$ \frac{dv}{dt} = -2u - v - 2uv - v^2 $ 

But since $ u, v $ are both very small (WHY IS THIS I DON'T GET WHY) then the non-linear terms are much smaller than the linear terms.

$ \frac{du}{dt} \approx - u - v $

$ \frac{dv}{dt} \approx -2u - v $ 

If you convert this to a matrix:

$ \left[ \begin{matrix} -1 & -1 \\ -2 & -1 \end{matrix} \right] \left[ \begin{matrix} u \\ v \end{matrix} \right]$

Then you could find the determinant using $ \lambda $, and using the characteristic equation, we find that the eigenvalues are:

$ \lambda = -1 \pm \sqrt{2} $

So this is unstable.

### Van der Pol equation

This is a non-linear system of equations. It has one equilibrium at $ (0,0) $

$ \frac{dx}{dt} = y $

$ \frac{dy}{dt} = -x + (1 - x^2) y $

There's no shifting of the equilibrium needed here, because the equilibrium is already at $ (0,0) $

So we can let $ x, y = 0.1 $

Then as in the previous example, the non-linear parts of the equation will be very small.

$ \frac{dx}{dt} = y $

$ \frac{dy}{dt} \approx -x + y $

![van der pol](Screenshot_2023-04-18_12-14-17.png)

You can see that it's a spiral source, so it's unstable, but you could find the eigenvalues to highlight the same fact.

$ \left[ \begin{matrix} 0 & 1 \\ -1 & 1 \end{matrix} \right] \left[ \begin{matrix} x \\ y \end{matrix} \right]$

Then again you end up with complex roots and they are complex eigenvalues, and this indicates it's unstable.
