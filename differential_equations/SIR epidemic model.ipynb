### SIR epidemic model

In this model, there are three kinds of people:

$ S $, susceptible to infenction

$ I $, infected and contagious

$ R $, recovered and no longer susceptible

These three can either be absolute values or you can imagine them as probabilities, and that means they are quotients that make the summation of 1.

Then the progression of a virus or bacterial outbreak etc is:

$ [S] \to [I] \to [R] $

The assumption is that a person enters the $ R $ state and remains there but there might be more complex models where a person can go back to $ S $.

$ \frac{dS}{dt} = -\alpha SI $

$ \frac{dI}{dt} = \alpha SI - \beta I $

$ \frac{dR}{dt} = \beta I $

If you have a think about the first equation, it is a product of variables $ S $ and $ I $. That's because those people need to interact before there's a transmition.

The final equation just has the one variable $ I $ and that's because infected people become recovered without any interaction from the other variables.

#### Random mixing

This is an assumption of the model: any infected person can transmit to any susceptible person.

#### Summation of derivatives

Recall that the summations of all population quotients is equal to 1.

$ S(t) + I(t) + R(t) = 1 $

If this is true then there is to be no net change in the population $ = 1 $. There might be a net change in reality as people die but here we consider the proportions and not absolute values.

Then the net growth of these proportions is zero:

$ \frac{dS}{dt} + \frac{dI}{dt} + \frac{dR}{dt} = 0 $

$ -\alpha SI + \alpha SI - \beta I + \beta I = 0 $

So, talking about absolute values would complicate things when all we want is proportions.

#### At the start of an outbreak

At the start the population of $ R = 0 $

$ S + I = 1 $

Then the outcome of a pandemic is completely determine by the rate of infection, in this model

![outcomes](Screenshot_2023-05-08_15-42-58.png)

Above the height of the curve indicates how many people become infected. The lower the $ \alpha $, the more people remain susceptible after the first wave of the pandemic, in the sense that we could model a single pandemic wave with this model.

Also, if $ \alpha $ is lower, then the derivative reaches 0 more quickly. In other words, the pandemic reaches its worst point sooner if the $ \alpha $ is lower.

#### Outbreak declines

Given the parameters $ \alpha, \beta $ but with varying initial conditions then there will be a threshold where the pandemic starts to decline.

If $ \frac{dI/dt}{dS/dt} = 0 $ then that is the point when the pandemic reaches no growth, neither negative or positive.

Then typically that's where the downward trend for the pandemic begins, where the number of susceptible people isn't enough to replenish those who exit infection by being recovered.

$ \frac{dI/dt}{dS/dt} = \frac{\alpha SI - \beta I}{-\alpha SI} $

$ \frac{dI/dt}{dS/dt} = \frac{\alpha S}{-\alpha S} - \frac{\beta }{-\alpha S} $

$ \frac{dI}{dS} = -1 + \frac{\beta }{\alpha} \frac{1}{S} $

And now, integrate both sides. This allows:

$ \frac{dI}{dS} = \int -1 + \frac{\beta }{\alpha} \frac{1}{S} dS $

$ I(S) =  -S + \frac{\beta }{\alpha} \ln S + c $

At $ t_0 $ we can assume that $ S \approx 1, I \approx 0 $

$ 0 =  -1 + \frac{\beta }{\alpha} \ln 1 + c $

$ c = 1 $

That means that with the assumed initial condition then the particular solution is:

$ I(S) = -S + \frac{\beta}{\alpha} \ln S + 1 $

#### Variant - births deaths

Let:

$ \mu = \frac{\text{births}}{\text{deaths}} $ be the growth rate of the population due to external causes

$ \beta $ be the transmition rate (formerly $ \alpha $)

$ \gamma $ be the recovery rate (formerly $ \beta $)

Don't forget, below, that $ 0 < S < 1 $

$ S' = \mu 1 - \mu S $ is the total change in the susceptible population due to natural causes, assuming that each person born is susceptible.

Then:

$ S' = \mu - \beta S I - \mu S $

$ I' = \beta S I - \mu I - \gamma I $

##### Stability analysis

The disease-free equilibrium is $ S^* = 1, I^* = 0 $

We can use a Jacobian Matrix to analyse this:

$ \left[ \begin{matrix} \frac{\partial S'}{\partial S} & \frac{\partial S'}{\partial I} \\ \frac{\partial I'}{\partial S} & \frac{\partial I'}{\partial I} \end{matrix} \right] = \left[ \begin{matrix} \beta I -\mu & \beta S \\ \beta I & \beta S - \mu -\gamma \end{matrix} \right] $

Substitute in the equilibria values: $ S^* = 1, I^* = 0 $

$ \left[ \begin{matrix} -\mu & \beta \\ 0 & \beta -\mu -\gamma \end{matrix} \right] \left[ \begin{matrix} S \\ I \end{matrix} \right] $

We could then multiply these two to get back the system of equations in linearised form around the equilibria, but what we really want is to know about the system's eigenvalues $ \lambda_1, \lambda_2 $, so we can do:

$ \left[ \begin{matrix} -\mu -\lambda & \beta \\ 0 & \beta -\mu -\gamma - \lambda \end{matrix} \right] $

FIND OUT HOW TO DERIVE THE EIGENVALUES

Then:

$ \lambda_1 = -\mu < 0 $, and this inequality is true because $ \mu > 0 $

$ \lambda_2 = \beta -\gamma - \mu $

Because $ \lambda_1 $ is negative then its eigenvalue will be a sink on the phase plane. The stability of the disease-free equilibrium now depends on $ \lambda_2 $. If $ \lambda_2 > 0 $ then this equilibrium is a saddle (with one positive and one negative eigenvalue). If $ \lambda_2 < 0 $. Then it is a stable sink.

But there is a threshold where $ \lambda_2 = 0 $, and it represents a scenario where the disease continues in the population but doesn't grow or decrease.

- Suppose $ \lambda_2 = 0 $:

$ \lambda_2 = \beta -\gamma -\mu = 0 $

$ \beta = \gamma +\mu  $

$ R_0 = \frac{\beta}{\gamma +\mu} = 1 $

The intuition here is to imagine that $ R_0 $ is how many people become infected per infected person.

- Suppose $ \lambda_2 < 0 $:

$ \lambda_2 = \beta -\gamma -\mu < 0 $

$ \beta < \gamma +\mu $

$ R_0 = \frac{\beta}{\gamma + \mu} < 1 $

This is a situation where the disease declines.

- Suppose $ \lambda_2 > 0 $:

$ \lambda_2 = \beta -\gamma -\mu > 0 $

$ \beta > \gamma +\mu $

$ R_0 = \frac{\beta}{\gamma + \mu} > 1 $

This is a situation where the disease spreads.

##### The other equilibrium - after the pandemic

After the pandemic $ S^* = 0, I^* = 0 $, because there are no more susceptible people left.

$ \left[ \begin{matrix} \frac{\partial S'}{\partial S} & \frac{\partial S'}{\partial I} \\ \frac{\partial I'}{\partial S} & \frac{\partial I'}{\partial I} \end{matrix} \right] = \left[ \begin{matrix} \beta I -\mu & \beta S \\ \beta I & \beta S - \mu -\gamma \end{matrix} \right] $

Substitute in the equilibria values: $ S^* = 0, I^* = 0 $

$ \left[ \begin{matrix} -\mu & 0 \\ 0 & -\mu -\gamma \end{matrix} \right] $

And now find the eigenvalues:

$ \left[ \begin{matrix} -\mu -\lambda & 0 \\ 0 & -\mu -\gamma -\lambda \end{matrix} \right] $

Using the characteristic eqution to find the determinant:

$ (-\mu -\lambda)(-\mu -\gamma -\lambda) = \mu^2 + \mu \gamma + \mu \lambda + \lambda \mu + \lambda \gamma + \lambda^2 $

$ (-\mu -\lambda)(-\mu -\gamma -\lambda) = \mu^2 + \mu (\gamma + \lambda) + \lambda (\mu + \gamma) + \lambda^2 $


$ -\mu \beta + \mu \gamma + \mu^2 + \mu \lambda - \lambda \beta + \lambda \gamma + \lambda \mu + \lambda^2 $

$ \mu (- \beta + \gamma + \mu + \lambda) + \lambda(- \beta + \gamma + \mu + \lambda) $
